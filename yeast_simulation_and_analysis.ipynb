####ver 1.4 - modify the time for intraribosomal events####

The effective kcat/km for ribosome trna association in Shah et al. 2013 is:

\begin{align} 
\frac{k_{cat}}{K_M} &= \frac{s}{N_t \tau_t}\\
                        &= \frac{7.78\times10^{-4}}{ (1.24\times10^{7}) \times (4.45\times10^{-7})} \\
                        &= 1.41 \times10^{-4} (\text{molecules per cell})^{-1}s^{-1} \\
                        &= 3.57 \times 10^{6} M^{-1}s^{-1}  \hskip1cm \text{(using the cell volume and Avogadro number)}\\
\end{align}

```python
%%bash
diff scripts/yeastsim1.1.c scripts/yeastsim1.4.c
```

## Create transgene sequences with different CAI ##

####Identify all ribosomal genes and write to Fasta file for creating CAI####

```python
revGeneticCode = CodonUsage.SynonymousCodons

ribosomalGenes = list()
# The ribosomal_genes_shah_2013.txt file is Table S3 in Shah et al. 2013
for line in open('seq_and_annotation/ribosomal_genes_shah_2013.txt').readlines():
    for word in line.split():
        if len(word) > 0:
            ribosomalGenes.append(word)

# fasta file below obtained from http://downloads.yeastgenome.org/sequence/S288C_reference/orf_dna/orf_coding.fasta.gz
seqFile = 'seq_and_annotation/orf_coding.fasta' 

ribosomalCds = list()
for record in SeqIO.parse( open( seqFile ), 'fasta' ):     
    # exclude these genes that have high sequencing counts but they have unknown function and skew the calculation of mRNA copy number.
    if record.id in ribosomalGenes:
        ribosomalCds.append( record )
        
revGeneticCode = CodonUsage.SynonymousCodons
SeqIO.write( ribosomalCds, "seq_and_annotation/ScRibosomalCds.fasta", "fasta");
```

####Generate 10 random GFP coding sequences for each CAI value that used in Shah et al. 2013####

```python
revGeneticCode = CodonUsage.SynonymousCodons

yeastCai = CodonUsage.CodonAdaptationIndex()
yeastCai.generate_index( 'seq_and_annotation/ScRibosomalCds.fasta' )

#Download GFP fasta protein sequence from uniprot: http://www.uniprot.org/uniprot/P42212
transgeneProteins = dict()
transgeneProteins[ 'gfp' ] = str( SeqIO.read( 'seq_and_annotation/GFP.fasta', 'fasta' ).seq )

# function starts with an input sequence and lowers cai by random synonymous mutations 
# till the desired cai target is reached.
def getSeqForTargetCai( inputSeq, targetCai ):
    currentSeq = inputSeq
    currentCai = yeastCai.cai_for_gene( currentSeq )
    seqLength = len(inputSeq) / 3
    numberOfIterations = 0
    while currentCai > targetCai:
        randPos = numpy.random.randint(0, seqLength )
        oldCodon = currentSeq[3*randPos:3*(randPos+1)]
        oldCodonCai = yeastCai.index[ str(oldCodon) ]
        aa = SeqUtils.seq3( Seq.Seq(oldCodon).translate() ).upper()
        newCodon = numpy.random.choice( revGeneticCode[ aa ] )
        newCodonCai = yeastCai.index[ newCodon ]
        if (newCodonCai < oldCodonCai and newCodonCai > 0):
            currentSeq = currentSeq[:3*randPos] + newCodon + currentSeq[3*(randPos+1):]            
        currentCai = yeastCai.cai_for_gene( currentSeq )
        numberOfIterations += 1
        if numberOfIterations%100 == 0:
            clear_output()
            print protein, targetCai, currentCai, numberOfIterations
            sys.stdout.flush()
    return currentSeq

#These values are from Table S4 in Shah 2013.
for caiValue in [0.093,0.164,0.222,0.303,0.377,0.461,0.542,0.589,0.607,
                 0.653,0.691,0.705,0.726,0.771,0.787,0.867,0.912,0.92,0.932,0.933,0.97]:    
    seqrecords = list()
    for protein in transgeneProteins:
        #find the nucleotide sequence with cai value of 1.
        maxCaiVariant = list()
        for aa in transgeneProteins[protein]:
            for codon in revGeneticCode[ SeqUtils.seq3(aa).upper() ]:
                if yeastCai.index[ codon ] == 1:
                    break
            maxCaiVariant.append(codon)
        maxCaiVariant = ''.join(maxCaiVariant)
        
        #create 10 random variants with the required cai value.
        for variant in range(10):
            newSeq = getSeqForTargetCai( maxCaiVariant, caiValue )
            fasta_id = protein + "_cai_" + "%0.3f"%caiValue + "_variant_" + str(variant+1)            
            seqrecords.append( SeqRecord( Seq.Seq( newSeq), id=fasta_id, description='') )
            
    outputFile = "seq_and_annotation/Sc.transgene.cai" + "%0.3f"%caiValue + ".fa"
    SeqIO.write( seqrecords, outputFile, "fasta" )
    
clear_output()
```

####Create input seq files with transgene sequences####

```python
mrnaFractionDict = {1:1, 10:10, 50:50}
for actualFraction,mrnaFraction in mrnaFractionDict.items():
    for cai in [0.093,0.164,0.222,0.303,0.377,0.461,0.542,0.589,0.607,
                 0.653,0.691,0.705,0.726,0.771,0.787,0.867,0.912,0.92,0.932,0.933,0.97]:
        outputFile = open( "seq_and_annotation/Sc.mrna.init.copynum.transgenefrac"
                          + str(actualFraction) + ".cai" + "%0.3f"%cai + ".seq", "write" )     
        totalNumberOfOldMrnas = 0
        initiationRateList = list()
        # The S.cer.genome is from Data S1 of Shah 2013
        for line in open( "seq_and_annotation/S.cer.genom" ).readlines():
            oldLine = line.split()
            oldMrnaNumber = int( oldLine[1] )
            initiationRateList.append( float( oldLine[0] ) )
            totalNumberOfOldMrnas += oldMrnaNumber
            outputFile.write(line)
        
        initationRate95percentile = numpy.percentile( initiationRateList, q=95 )
        numberOfTransgeneSequences = sum(map(lambda x: 1, SeqIO.parse( "seq_and_annotation/Sc.transgene.cai" + "%0.3f"%cai + ".fa", 'fasta' )))
        transgeneMrnaCopyNumber = numpy.ceil( (totalNumberOfOldMrnas * mrnaFraction / 100.0)
                                             / float(numberOfTransgeneSequences) )
        for record in SeqIO.parse( "seq_and_annotation/Sc.transgene.cai" + "%0.3f"%cai + ".fa", 'fasta' ):
            outputFile.write( '%0.7f'%initationRate95percentile + ' ' +
                                  str(int(transgeneMrnaCopyNumber)) + ' ' )
                                  
            for loop in range( 0, len(record)-6, 3 ):
                outputFile.write( str( geneticCode.ix[ str(record.seq[loop:loop+3] )]['index'] ) + ' ' )
            
            outputFile.write( str( geneticCode.ix[ str(record.seq[-6:-3] )]['index'] ) + '\n' ) 
        outputFile.close()        
```

##Running the simulation##

Steps for compiling the source are identical to those used in Shah et al. 2013 (see also simulation.ipynb).

The seq files from transgenes generated above have to be moved to example/input (in the Shah et al. Data S1 SMoPT folder hierarchy which is downloaded to the current working directory) before you run the following simulations.

```python
%%writefile scripts/run_shah_simulation.py

import sys
import subprocess
import numpy

jobindex = 0

for ribosomes in range(20000,200001,20000):
    for starvationLevel in [1,2,5,10]:
        jobindex += 1
        if sys.argv[1] == str(jobindex):
            subprocess.call( [ "scripts/yeastsim1.2 " + 
                               "-T 1500 " + 
                               "-H 1000 " + 
                               "-R " + str(ribosomes) +
                               " -t 3300000 " + 
                               "-N 3795 " +
                               "-L " + str(starvationLevel) + 
                               " -F example/input/S.cer.genom " + 
                               "-C example/input/S.cer.tRNA " +
                               "-s 1413 "
                               " -O output_files/yeastsim1.2/output_leucine_" + str(starvationLevel) + 
                               "_ribo" + str(ribosomes) +
                               " -p1 -p2 -p3 -p4 -p5" ], shell = True )
print jobindex

for ribosomes in range(20000,200001,20000):
    for starvationLevel in [1,5,20,100]:
        jobindex += 1
        if sys.argv[1] == str(jobindex):
            subprocess.call( [ "scripts/yeastsim1.3 " + 
                               "-T 1500 " + 
                               "-H 1000 " + 
                               "-R " + str(ribosomes) +
                               " -t 3300000 " + 
                               "-N 3795 " +
                               "-L " + str(starvationLevel) + 
                               " -F example/input/S.cer.genom " + 
                               "-C example/input/S.cer.tRNA " +
                               "-s 1413 "
                               " -O output_files/yeastsim1.3/output_leucine_" + str(starvationLevel) + 
                               "_ribo" + str(ribosomes) +
                               " -p1 -p2 -p3 -p4 -p5" ], shell = True )
print jobindex


# Transgene overexpression simulations with two different times for intra-ribosomal events.
for mrnaFraction in map( str, [1, 10, 50] ):
    for cai in map( lambda x: "%0.3f"%x, [0.093,0.164,0.222,0.303,0.377,0.461,0.542,0.589,0.607,
                             0.653,0.691,0.705,0.726,0.771,0.787,0.867,0.912,0.92,0.932,0.933,0.97] ):
        for (ribotrnakcatkm, codonminelongtime) in zip( ["3.56e7","3.56e6"], ["0.10", "0.00"] ):
            jobindex += 1
            if sys.argv[1] == str(jobindex):
                subprocess.call( [ "scripts/yeastsim1.4 " + 
                                   "-T 1500 " + 
                                   "-H 1000 " + 
                                   "-R 200000 " + 
                                   "-t 3300000 " + 
                                   "-N 3805 " + 
                                   "-F example/input/transgeneinput/Sc.mrna.init.copynum.transgenefrac" +
                                   mrnaFraction + ".cai" + cai + ".seq" +
                                   " -C example/input/S.cer.tRNA " + 
                                   "-s 1413 " + 
                                   "-O output_files/yeastsim1.4/output_cai_" + cai + "_mrnafraction_" + mrnaFraction +
                                   " -p1 -p2 -p3 -p4 -p5 " + 
                                   "--ribotrnakcatkm " + ribotrnakcatkm + 
                                   " --min-elong-time " + codonminelongtime ], shell= True )
print jobindex
```

Total of 212 simulatons for above set of parameters.

Each simulation run takes ~3 hours on a 32-bit single core laptop.

For fast run times, run each iteration on a separate node of a cluster. Alternately reduce the run time above from 1500s to 150s and threshold time from 1000s to 100s.

```python
%%bash

for jobindex in {1..212};
    do
    python scripts/run_shah_simulation.py $jobindex;
    done;
```

```python
allFiles = !ls output_files/yeastsim1.4/output*_gene_totetimes.out
markerfacecolor = {'0.00': 'white', '0.10': 'grey' }
markerList = {'1': 's', '10': 'o', '50': '^', }

inputseqfiles = !ls seq_and_annotation/Sc.mrna.init.copynum.transgenefrac*.seq

def plot_transgeneproteins_varyingcai():
    
    fig = plt.figure()
    fig.set_size_inches((5,5))  
    common_ax = plt.axes(frameon=False)
    common_ax.set_xticks([])
    common_ax.set_yticks([])
    common_ax.yaxis.labelpad = 50
    common_ax.xaxis.labelpad = 40   
    
    axisIndex = 0  
    ax = fig.add_subplot(2,2,axisIndex+1)
    
    for codonelongtime in ['0.00','0.10']:

        mrnaFractionValues = numpy.unique( 
                    [re.search( 'mrnafraction_([\d\.]+)_', x ).groups(0)[0] for x in allFiles ] ) 

        for mrnafraction in mrnaFractionValues:
            totalNumberOfProteins = dict()
            cai = dict()
            sampleIndex = 0
            subsetFiles = filter( lambda x: x.find( 'minelongtime_' + codonelongtime + '_' ) != -1, allFiles )
            subsetFiles = filter( lambda x: x.find( 'mrnafraction_' + mrnafraction + '_' ) != -1, subsetFiles )
            for File in subsetFiles: 
                currentcai = re.search( 'cai_([\d\.]+)_', File ).groups(0)[0]
                cai[ sampleIndex ] = float( currentcai ) 
                
                totalNumberOfProteins[ sampleIndex ] = 0            
                proteinsProduced = 0
                mrnaNumber = 0
                for line in open( File ).readlines()[1:]: 
                    geneId = int( line.split()[0] )
                    if geneId in range(3795,3805):
                        proteinsProduced += float( line.split()[1] )
                                    
                inputseqfile = filter(lambda x: x.find( 'transgenefrac' + mrnafraction + "." ) != -1 and
                                      x.find( 'cai' + currentcai ) != -1,
                                      inputseqfiles
                                      )[0]
                        
                for linenumber,line in enumerate(open( inputseqfile ).readlines()):                     
                    if linenumber in range(3794,3804):
                        mrnaNumber += float( line.split()[1] )            

                totalNumberOfProteins[ sampleIndex ] =   proteinsProduced / float( mrnaNumber )
                
                sampleIndex += 1

            caiArray = numpy.zeros( (len(cai),), dtype=float )    

            for index, key in enumerate( sorted( cai, key = cai.get ) ):                
                caiArray[ index ] = cai[ key ] 

            totalNumberOfProteinsArray = numpy.zeros( (len(totalNumberOfProteins),), dtype=float)    
            for index, key in enumerate( sorted( cai, key = cai.get ) ):
                totalNumberOfProteinsArray[ index ] = totalNumberOfProteins[ key ]

            ax.plot( caiArray, totalNumberOfProteinsArray / totalNumberOfProteinsArray.max(),
                                    '-', marker = markerList[mrnafraction], label = mrnafraction + "%, " + codonelongtime[:-1] + "s", markersize=6,
                markerfacecolor = markerfacecolor[codonelongtime], markeredgecolor = 'black', color = 'black' )


        spineOffset = {'left': 5, 'bottom': 0}    
        [ spine.set_position(('outward',spineOffset[loc])) if loc in ['left','bottom'] else spine.set_color('none') for loc, spine in ax.spines.items() ]         

        ax.set_xlim(left = 0, right = 1)
        ax.set_ylim(bottom = 0.4, top = 1.05 )
        ax.xaxis.set( ticks_position = 'bottom', ticks=numpy.arange(0,1.1,0.3) )
        ax.yaxis.set( ticks_position = 'left', ticks=numpy.arange(0.4,1.1,0.2) )                    
        ax.legend(loc='lower right', title=u'Transgene fraction\nof transcriptome,τ0', bbox_to_anchor = (1.1,0,1,1))
        ax.set_xlabel( 'Codon Adaptation Index' )
        ax.set_ylabel( 'Transgene proteins / mRNA\n(relative to max)' ) 
        
    fig.savefig('transgene_proteins_vs_cai_yeast.svg')       
    
plot_transgeneproteins_varyingcai();
```

## Figures S2F, S2G ##

```python
markerList = dict()
markerList['1.2'] = {'1': 'o', '2': 's', '5': 'd', '10': '*' }
markerList['1.3'] = {'1': 'o', '5': 's', '20': 'd', '100': '*' }
           
markerColor = dict()
markerColor['1.2'] = {'1': 'blue', '2': 'green', '5': 'orange', '10': 'red' }
markerColor['1.3'] = {'1': 'blue', '5': 'green', '20': 'orange', '100': 'red' }

legendTitle = dict()
legendTitle['1.2'] = "Total tRNA fold-change"
legendTitle['1.3'] = "Charged tRNA fold-change"

lengths = list()
for line in open('seq_and_annotation/S.cer.genom').readlines():
    lengths.append( len(line.split()[2:]) )

def plot_totalproteins_uniformtotaltrna( ): 
    
    fig = plt.figure()
    fig.set_size_inches((5,5))     
    common_ax = plt.axes(frameon=False)
    common_ax.set_xticks([])
    common_ax.set_yticks([])
    common_ax.yaxis.labelpad = 50
    common_ax.xaxis.labelpad = 40
    
    axisCount = 0
    for aaCount, aminoacid in enumerate( ['leucine'] ):
        for version in ['1.2','1.3']:
            allFiles = os.listdir(os.getcwdu() + '/output_files/yeastsim' + version )            
            allFiles = filter( lambda x: x.find( 'gene_totetimes') != -1, allFiles)            
            allFiles = filter( lambda x: x.find( aminoacid ) != -1, allFiles)
            
            starvationindices = numpy.unique( [ int( re.search( aminoacid + '_(\d+)_', x ).groups(0)[0] ) for x in allFiles ] )
            if version == '1.3':
                starvationindices = starvationindices[::2]
            ax = fig.add_subplot(2,2, axisCount+1)        
            for starvationindex in starvationindices:
                totalNumberOfProteins = dict()
                numberOfRibosomes = dict()
                sampleIndex = 0
                subsetFiles = filter( lambda x: x.find( aminoacid + '_' + str(starvationindex) + '_' ) != -1, allFiles )
                subsetFiles = sorted( subsetFiles, key= lambda x: int( re.search( '_ribo(\d+)_', x ).groups()[0] ), reverse=False )
                for File in subsetFiles: 

                    totalNumberOfProteins[ sampleIndex ] = 0   
                    lineNumber = 0
                    for line in open( 'output_files/yeastsim' + version + '/' + File ).readlines()[1:]:                        
                        totalNumberOfProteins[ sampleIndex ] += float( line.split()[1] ) * lengths[lineNumber]
                        lineNumber += 1

                    numberOfRibosomes[ sampleIndex ] = int( re.search( '_ribo(\d+)_', File ).groups()[0] )            

                    sampleIndex += 1

                numberOfRibosomesArray = numpy.zeros( (len(numberOfRibosomes),), dtype=int32 )    
                for key in numberOfRibosomes:
                    numberOfRibosomesArray[ key ] = numberOfRibosomes[ key ]

                totalNumberOfProteinsArray = numpy.zeros( (len(totalNumberOfProteins),), dtype=float)    
                for key in totalNumberOfProteins:
                    totalNumberOfProteinsArray[ key ] = totalNumberOfProteins[ key ]

                if starvationindex == 1:
                    maxProtein = numpy.max( totalNumberOfProteinsArray )

                ax.plot( numberOfRibosomesArray / float( max(numberOfRibosomesArray) ), 
                totalNumberOfProteinsArray/maxProtein, 
                color = markerColor[version][ str(starvationindex) ],
                markerfacecolor = markerColor[version][ str(starvationindex) ],
                markeredgecolor = markerColor[version][ str(starvationindex) ],
                marker = markerList[version][ str(starvationindex) ],
                markersize=4,
                label = str(starvationindex) + '-fold'    )

            spineOffset = {'left': 5, 'bottom': 0}    
            [ spine.set_position(('outward',spineOffset[loc])) if loc in ['left','bottom'] else spine.set_color('none') for loc, spine in ax.spines.items() ]         
            ax.set_xlim(left = 0, right = 1.1)
            ax.set_ylim(bottom = -0.05, top = 1.05 )
            ax.xaxis.set( ticks_position = 'bottom', ticks=numpy.arange(0.1,1.1,0.3) )
            ax.yaxis.set( ticks_position = 'left', ticks=numpy.arange(0.1,1.1,0.3) )            
            ax.set_ylabel( '' )
            ax.legend( loc='center left', title=legendTitle[version], bbox_to_anchor = (0,-1.2,1,1))
            ax.set_title( aminoacid + ' starvation')
            if axisCount == 0:
                ax.set( ylabel = 'Total Protein Synthesis Rate\n(Relative to Wild-type)' )
            ax.set( xlabel = 'Number of Ribosomes\n(Relative to Wild-type)' )
            axisCount += 1
            
    fig.savefig( 'pss_vs_ribo_yeast.svg' )
    
plot_totalproteins_uniformtotaltrna() 
```
