#### Create input files for tRNAs and codons data ####
There are four files for:
 * tRNA concentration
 * aminoacylation rate for each tRNA during rich growth
 * cognate codon-tRNA pairs
 * codon-trna interaction weights for cognate codon-tRNA pairs 

````python
# Read tRNA concentrations, anticodon, cognate codons 
# from Table 2 in Dong et al. J. Mol. Biol. 1996.
lines = open( 'seq_and_annotation/Ecoli.trna.abundance.dong1996.txt'
             ).read().splitlines()
trnas = dict()
#read every 6th line until last but one.
for trnaindex, trna in enumerate( lines[0:-1:6] ): 
    trnas[ trna ] = dict()
    trnas[ trna ]['anticodon'] = lines[trnaindex*6 + 1
                                       ].strip().replace('U','T')
    trnas[ trna ]['codons'] = [codon.strip().replace('U','T')
                               for codon in lines[trnaindex*6 + 2
                                                  ].split( ',' )]    
    trnas[ trna ]['mean_concn'] = int( lines[trnaindex*6 + 4].split('(')[0] )
    trnas[ trna ]['std_concn'] = int( lines[trnaindex*6 + 4
                                            ].split('(')[1].split(')'
                                                            )[0].strip() )
    trnas[ trna ]['fraction'] = float( lines[trnaindex*6 + 5].strip() )    
    trnas[ trna ]['aminoacid'] = trna[:3].lower()

# Ile3 was not measured in Dong 1996. So assigning it to Ile1.    
trnas[ 'Ile1' ]['codons'].append( 'ATA' ) 
# Leu3 was found to read CTT in Sørensen et al. J. Mol. Biol. 2005.
trnas[ 'Leu3' ]['codons'].append( 'CTT' )

# Calculate the factors for codon-tRNA interaction 
# that multiply the kcat/KM for codon-tRNA reading:
# w = 1 for Watson-Crick base pairs,
# w = 0.64 for purine-pyrimidine mismatch,
# w = 0.61 for purine-purine mismatch. (Weights based on Shah 2013).

# If two trnas have the same anticodon 
# they are considered as a single trna species in the simulation.
concentrations = dict() 
cognate_pairings = dict()
weights = dict()
# For Fig. S1A, weighted sum of all cognate trna concentrations for each codon 
trna_abundance = dict() 

for trna in trnas:    
    cognatecodons = trnas[trna]['codons']
    if trnas[trna]['anticodon'] not in concentrations:
        concentrations[trnas[trna]['anticodon']] = trnas[trna]['mean_concn']
    else:
        concentrations[trnas[trna]['anticodon']] += trnas[trna]['mean_concn']
        
    for codon in cognatecodons: 
        # pairing at the wobble position.
        pairing = codon[-1:] + trnas[trna]['anticodon'][:1] 
        watson_crick_pairs = ['GC','AT','TA','CG']
        pur_pyr_pairs = ['GT','TG','CA']
        pur_pur_pairs = ['TT','AA']
        if pairing in watson_crick_pairs:
            weight = 1
        elif pairing in pur_pyr_pairs:
            weight = 0.64
        else:
            weight = 0.61
            
        # Cognate tRNA assignment below based on data in Sørensen et al. J. Mol. Biol. 2005.    
        if codon in ['CTT'] and trnas[trna]['anticodon'] in ['TAG']:
            weight = 1.0 
            
        if codon not in cognate_pairings:
            cognate_pairings[ codon ] = [ trnas[trna]['anticodon'] ]
            weights[ codon ] = [ weight ]            
        elif trnas[trna]['anticodon'] not in cognate_pairings[ codon ]:
            cognate_pairings[ codon ].append( trnas[trna]['anticodon'] )
            weights[ codon ].append( weight )
            
        if codon.replace('U','T') not in trna_abundance.keys():
            trna_abundance[ codon.replace('U','T') ] = int( 
                        trnas[trna]['mean_concn'] * weight )
        else:
            trna_abundance[ codon.replace('U','T') ] = int(
                        trna_abundance[ codon.replace('U','T') ] 
                                + trnas[trna]['mean_concn'] * weight )
            
codon_parameters = pandas.DataFrame.from_dict( 
                            { 'trna_weighted_concn' : trna_abundance } )
codon_parameters[['trna_weighted_concn']].to_csv(
    'seq_and_annotation/Ecoli.trna.concn.for.codon.csv', index='codon')

#trnaindices are sorted alphabetically.

# concentrations file.            
outputFile = open( "seq_and_annotation/Ecoli.trna.concentrations.tsv"
                  , "write" )
outputFile.write( 'trnaindex\tanticodon\tmolpercell\taminoacid\n' )
# trnaindex 0 is a dummy trna without any anticodon.
outputFile.write( '0\tNNN\t0\txxx\n' ) 
trnaindices = dict()
for trnaindex, anticodon in enumerate( sorted(concentrations) ):
    aminoacid = [ trnas[trna]['aminoacid'] 
                 for trna in trnas 
                 if trnas[trna]['anticodon'] == anticodon ][0]
    trnaindices[ anticodon ] = trnaindex + 1
    outputFile.write( str(trnaindex + 1) + '\t' 
                     + anticodon + '\t' + '%d'%concentrations[anticodon] 
                     + '\t' + aminoacid + '\n' )    
outputFile.close()  

# aminoacylation rates file.
outputFile = open( "seq_and_annotation/Ecoli.trna.aminoacylation.rate.tsv"
                  , "write" )
outputFile.write( 'trnaindex\tanticodon\tkcatBykm\n' )
outputFile.write( '0\tNNN\t0\n' )
trnaindices = dict()
for trnaindex, anticodon in enumerate( sorted(concentrations) ):
    trnaindices[ anticodon ] = trnaindex + 1
    outputFile.write( str(trnaindex + 1) + '\t'
                     + anticodon + '\t' 
                     + '%d'% fixed_parameters['max_aminoacylation_rate'] 
                     + '\n' )    
outputFile.close() 

# cognate pairings file.
outputFile = open( "seq_and_annotation/Ecoli.cognate.pairs.tsv", "write" )
outputFile.write( 'codonindex\tcodon\ttrnaindex1\ttrnaindex2\n' )
for codon in geneticCode.index:
    if len( cognate_pairings[codon] ) == 1:    
        outputFile.write( str(geneticCode.ix[codon]['index']) 
                     + '\t' + codon + '\t' 
                     + '%d'%trnaindices[cognate_pairings[codon][0]] 
                     + '\t0\n' )
    elif len( cognate_pairings[codon] ) == 2:    
        outputFile.write( str(geneticCode.ix[codon]['index']) 
                     + '\t' + codon                           
                     + '\t' + '%d'%trnaindices[cognate_pairings[codon][0]] 
                     + '\t' + '%d'%trnaindices[cognate_pairings[codon][1]] 
                     + '\n' )
outputFile.close() 

# cognate weights file.
outputFile = open( "seq_and_annotation/Ecoli.cognate.weights.tsv", "write" )
outputFile.write( 'codonindex\tcodon\ttrnaweight1\ttrnaweight2\n' )
for codon in geneticCode.index:
    if len( weights[codon] ) == 1:    
        outputFile.write( str(geneticCode.ix[codon]['index']) 
                         + '\t' + codon + '\t' 
                         + '%0.2f'%weights[codon][0] + '\t0.00\n' )
    elif len( weights[codon] ) == 2:    
        outputFile.write( str(geneticCode.ix[codon]['index'])
                         + '\t' + codon + 
                         '\t' + '%0.2f'%weights[codon][0] 
                         + '\t' + '%0.2f'%weights[codon][1] + '\n' )
outputFile.close() 
````

#### Modify the codon-tRNA interaction rates to use the exact measurement values from Sorensen et al. J. Mol. Biol. 2005. ####

````python
# Read tRNA concentrations, anticodon, cognate codons 
# from Table 2 in Dong et al. J. Mol. Biol. 1996.
lines = open( 'seq_and_annotation/Ecoli.trna.abundance.dong1996.txt'
             ).read().splitlines()
trnas = dict()
#read every 6th line until last but one.
for trnaindex, trna in enumerate( lines[0:-1:6] ): 
    trnas[ trna ] = dict()
    trnas[ trna ]['anticodon'] = lines[trnaindex*6 + 1
                                       ].strip().replace('U','T')
    trnas[ trna ]['codons'] = [codon.strip().replace('U','T')
                               for codon in lines[trnaindex*6 + 2
                                                  ].split( ',' )]    
    trnas[ trna ]['mean_concn'] = int( lines[trnaindex*6 + 4].split('(')[0] )
    trnas[ trna ]['std_concn'] = int( lines[trnaindex*6 + 4
                                            ].split('(')[1].split(')'
                                                            )[0].strip() )
    trnas[ trna ]['fraction'] = float( lines[trnaindex*6 + 5].strip() )    
    trnas[ trna ]['aminoacid'] = trna[:3].lower()

# Ile3 was not measured in Dong 1996. So assigning it to Ile1.    
trnas[ 'Ile1' ]['codons'].append( 'ATA' ) 
# Leu3 was found to read CTT in Sørensen et al. J. Mol. Biol. 2005.
trnas[ 'Leu3' ]['codons'].append( 'CTT' )

# Calculate the factors for codon-tRNA interaction 
# that multiply the kcat/KM for codon-tRNA reading:
# w = 1 for Watson-Crick base pairs,
# w = 0.64 for purine-pyrimidine mismatch,
# w = 0.61 for purine-purine mismatch. (Weights based on Shah 2013).

# If two trnas have the same anticodon 
# they are considered as a single trna species in the simulation.
concentrations = dict() 
cognate_pairings = dict()
weights = dict()
# For Fig. S1A, weighted sum of all cognate trna concentrations for each codon 
trna_abundance = dict() 

for trna in trnas:    
    cognatecodons = trnas[trna]['codons']
    if trnas[trna]['anticodon'] not in concentrations:
        concentrations[trnas[trna]['anticodon']] = trnas[trna]['mean_concn']
    else:
        concentrations[trnas[trna]['anticodon']] += trnas[trna]['mean_concn']
        
    for codon in cognatecodons: 
        # pairing at the wobble position.
        pairing = codon[-1:] + trnas[trna]['anticodon'][:1] 
        watson_crick_pairs = ['GC','AT','TA','CG']
        pur_pyr_pairs = ['GT','TG','CA']
        pur_pur_pairs = ['TT','AA']
        if pairing in watson_crick_pairs:
            weight = 1
        elif pairing in pur_pyr_pairs:
            weight = 0.64
        else:
            weight = 0.61
            
        # Based on data in Sørensen et al. J. Mol. Biol. 2005.
        # Default weight for all other codons is 1 (kel = 2x10^7 M-1s-1).
        if codon in ['CTG']:
            if trnas[trna]['anticodon'] in ['CAG']:
                weight = 18.2 / 20.0 
            if trnas[trna]['anticodon'] in ['TAG']:
                weight = 9.6 / 20.0 
        if codon in ['CTC']:
            if trnas[trna]['anticodon'] in ['GAG']:
                weight = 30.3 / 20.0 
        if codon in ['CTT']:
            if trnas[trna]['anticodon'] in ['TAG']:
                weight = 11.4 / 20.0 
            if trnas[trna]['anticodon'] in ['GAG']:
                weight = 32.8 / 20.0 
        if codon in ['CTA']:
            if trnas[trna]['anticodon'] in ['TAG']:
                weight = 11.6 / 20.0
            
        if codon not in cognate_pairings:
            cognate_pairings[ codon ] = [ trnas[trna]['anticodon'] ]
            weights[ codon ] = [ weight ]            
        elif trnas[trna]['anticodon'] not in cognate_pairings[ codon ]:
            cognate_pairings[ codon ].append( trnas[trna]['anticodon'] )
            weights[ codon ].append( weight )
            
        if codon.replace('U','T') not in trna_abundance.keys():
            trna_abundance[ codon.replace('U','T') ] = int( 
                        trnas[trna]['mean_concn'] * weight )
        else:
            trna_abundance[ codon.replace('U','T') ] = int(
                        trna_abundance[ codon.replace('U','T') ] 
                                + trnas[trna]['mean_concn'] * weight )

# cognate weights file.
outputFile = open( "seq_and_annotation/Ecoli.cognate.weights.sorensen.tsv", "write" )
outputFile.write( 'codonindex\tcodon\ttrnaweight1\ttrnaweight2\n' )
for codon in geneticCode.index:
    if len( weights[codon] ) == 1:    
        outputFile.write( str(geneticCode.ix[codon]['index']) 
                         + '\t' + codon + '\t' 
                         + '%0.2f'%weights[codon][0] + '\t0.00\n' )
    elif len( weights[codon] ) == 2:    
        outputFile.write( str(geneticCode.ix[codon]['index'])
                         + '\t' + codon + 
                         '\t' + '%0.2f'%weights[codon][0] 
                         + '\t' + '%0.2f'%weights[codon][1] + '\n' )
outputFile.close() 
````

#### Parse *E. coli* genome sequence and coding sequence annotations to create a fasta file of all coding sequences ####

````python
# .fna and .ptt files below downloaded from NCBI on 19th Feb 2014
genome = Bio.SeqIO.read( 'seq_and_annotation/NC_000913.fna', 'fasta' )
geneTable = pandas.read_table( 'seq_and_annotation/NC_000913.ptt',
                              skiprows=2 )
# Parse start coordinate of each CDS.
geneTable['Start'] = geneTable['Location'].apply( 
                    lambda x: numpy.int32( x.split('.')[0] ) ) 
# Part end coordinate of each CDS.
geneTable['End'] = geneTable['Location'].apply( 
                    lambda x: numpy.int32( x.split('.')[-1] ) )

outputFile = open( 'seq_and_annotation/Ecoli.orfs.fasta', 'write' )
genes = list()
for row in geneTable.iterrows():
    gene = row[1]['Gene']
    if gene in genes:
        gene = gene + row[1]['Synonym']
    # The length of these gene sequences is not a multiple of 3, 
    # or contains stop codons, and are excluded from simulation.
    if re.search( 'prfB|nrdF|fdnG|fdoG|fdhF', gene ): 
        continue 
    #1-based .ptt file indexing to 0-based python indexing.
    start = row[1]['Start'] - 1 
    # no -1 necessary since end-point is included in .ptt but not in python.
    stop = row[1]['End'] 
    strand = row[1]['Strand']
    seq = genome[ start:stop ].seq
    # reverse_complement if on reverse strand.
    if strand == '-': 
        seq = seq.reverse_complement()        
    outputFile.write( '>' + gene + '\n' + str(seq) + '\n' ) 
    genes.append( gene )
outputFile.close()
````

#### Calculate initiation rates and mRNA copy numbers for each coding sequence, and write it to an outputfile that also includes the numeric codon sequence. ####

````python
samples = ['wt_yesleu_mono', 'wt_yesleu_total', 'wt_noleu_mono', 'wt_noleu_total']

df = list()

for sample in samples:    
    temp = pandas.DataFrame.from_dict(  pickle.load( open( 'output_files/sequencing_output/geneStats_'+sample+'.pickled', 'rb' ) ), orient='index' )                        
    df.append( temp.drop( ['meanReadDensity'], axis=1 ).rename( columns = {'totalReads': 'counts_' + sample} ) )        
    
geneTable = pandas.read_table( 'seq_and_annotation/NC_000913.ptt', skiprows=2 )
geneTable['Start'] = geneTable['Location'].apply( lambda x: numpy.int32( x.split('.')[0] ) ) # Parse start coordinate of each CDS.
geneTable['End'] = geneTable['Location'].apply( lambda x: numpy.int32( x.split('.')[-1] ) ) #Part end coordinate of each CDS.

genes = dict()
for row in geneTable.iterrows():
    gene = row[1]['Gene']
    if gene in genes:
        gene = gene + row[1]['Synonym']
    # The length of these gene sequences is not a multiple of 3, 
    # or contains stop codons, and are excluded from simulation.
    if re.search( 'prfB|nrdF|fdnG|fdoG|fdhF', gene ): 
        continue 
    genes[ gene ] = {
                     'function': row[1]['Product'],
                     'strand':  row[1]['Strand'],
                     'start': row[1]['Start'] - 1,
                     'stop': row[1]['End'],
                     'length': row[1]['End'] - row[1]['Start'] + 1 - 3 # exclude stop codon in length
                     }
df.append( pandas.DataFrame.from_dict(genes, orient='index') )
        
data = pandas.concat( df, join='inner', axis=1 )

for col in data:
    if col.startswith( 'counts' ):
        data[col] = data[col].apply( numpy.float )        
        data[ col.replace('counts','lengthnorm') ] =  numpy.divide( data[ col ], data[ 'length' ] )
        if col.endswith('total'):
            cellNormalization = sum( data[ col ] ) / fixed_parameters['mean number of mrnas']            
        elif col.endswith('mono'):
            cellNormalization = sum( data[ col ] ) / ( fixed_parameters['mean number of ribosomes']
                                                    * fixed_parameters['fraction of mrna bound ribosomes'] )
        
        data[ col.replace('counts','cellnorm') ] = data[ col ].apply( lambda x: x  / cellNormalization )        
        
initiationRateList = []
for row in data.iterrows():
    # exclude genes that do not pass this read density threshold.
    if not row[1]['lengthnorm_wt_noleu_mono'] > 1.0 / 3.0:
        continue
    # if genes get > 10 reads in ribosome profiling and rna seq, infer initiation rate. If not, assign
    # the median initiation rate.
    if row[1]['counts_wt_yesleu_mono'] > 10 and row[1]['counts_wt_yesleu_total'] > 10:
        initiationRateList.append( row[1]['cellnorm_wt_yesleu_mono']
                                  / row[1]['cellnorm_wt_yesleu_total']
                                  * fixed_parameters['mean normal elongation rate'] )

medianInitiationRate = numpy.median( initiationRateList )

outputFile = open( 'seq_and_annotation/init_rates_and_mrna_numbers.csv', 'write' )
outputFile.write( 'gene\tinitiation rate (sec)\tmRNA copy number (per cell)\tgene class\n' )
totalNumberOfMrnas = 0
geneCount = 0
for row in data.iterrows():
    # exclude genes that do not pass this read density threshold.
    if not row[1]['lengthnorm_wt_noleu_mono'] > 1.0 / 3.0:
        continue
    if row[1]['counts_wt_yesleu_mono'] > 10 and row[1]['counts_wt_yesleu_total'] > 10:
        initiationRate = ( row[1]['cellnorm_wt_yesleu_mono'] 
                            / row[1]['cellnorm_wt_yesleu_total'] 
                            * fixed_parameters['mean normal elongation rate'] )
    else:
        initiationRate = medianInitiationRate
        
    gene = row[0]

    # If inferred mRNA copy number is less than 1, change it to 1.
    if row[1]['cellnorm_wt_yesleu_total']/ row[1]['length'] > 1.0:
        mrnaCopyNumber = numpy.ceil( row[1]['cellnorm_wt_yesleu_total'] / row[1]['length'] )
    else:
        mrnaCopyNumber = 1
        
    outputFile.write( '%s\t%0.6f\t%d\n'%(row[0],initiationRate,mrnaCopyNumber) )
    geneCount += 1
    totalNumberOfMrnas += mrnaCopyNumber

outputFile.close()


inputFile = 'seq_and_annotation/init_rates_and_mrna_numbers.csv'
initRateAndMrnaCopyNumbers = pandas.read_table( inputFile, index_col=0 )
outputFile = open( 'seq_and_annotation/Ecoli.input.seq', 'write' )
record_count = 0
for record in Bio.SeqIO.parse( 'seq_and_annotation/Ecoli.orfs.fasta', 'fasta' ):
    if record.id in initRateAndMrnaCopyNumbers.index:
        outputFile.write( '%0.7f'%initRateAndMrnaCopyNumbers.ix[record.id]['initiation rate (sec)'] + ' ' +
                              str(int(initRateAndMrnaCopyNumbers.ix[record.id]['mRNA copy number (per cell)'])) + ' ' )
        for loop in range( 0, len(record)-6, 3 ):
            outputFile.write( str( geneticCode.ix[ str(record.seq[loop:loop+3] )]['index'] ) + ' ' )
        outputFile.write( str( geneticCode.ix[ str(record.seq[-6:-3] )]['index'] ) + '\n' )            
        record_count += 1
outputFile.close()
````

#### Create transgene sequences with different CAI ####

````python
sharpEcoliIndex = CodonUsageIndices.SharpEcoliIndex
cognateCodons = CodonUsage.SynonymousCodons
sharpEcoliCai = CodonUsage.CodonAdaptationIndex()

# function starts with an input sequence and lowers cai by random synonymous mutations 
# till the desired cai target is reached.
def getSeqForTargetCai( inputSeq, targetCai ):
    currentSeq = inputSeq
    currentCai = sharpEcoliCai.cai_for_gene( currentSeq )
    seqLength = len(inputSeq) / 3
    numberOfIterations = 0
    while currentCai > targetCai and numberOfIterations < 10000:
        randPos = numpy.random.randint(0, seqLength )
        oldCodon = currentSeq[3*randPos:3*(randPos+1)]
        oldCodonCai = sharpEcoliIndex[ str(oldCodon) ]
        aa = Bio.SeqUtils.seq3( Bio.Seq.Seq(oldCodon).translate() ).upper()
        newCodon = numpy.random.choice( cognateCodons[ aa ] )
        newCodonCai = sharpEcoliIndex[ newCodon ]
        if (newCodonCai < oldCodonCai):
            currentSeq = currentSeq[:3*randPos] + newCodon + currentSeq[3*(randPos+1):]            
        currentCai = sharpEcoliCai.cai_for_gene( currentSeq )
        numberOfIterations += 1
        if numberOfIterations%100 == 0:
            clear_output()
            print protein, targetCai, currentCai, numberOfIterations
            sys.stdout.flush()
    return currentSeq


ncbiproteins = {'insulin' : 'NP_000198.1',}

transgeneProteins = dict()

for protein in ncbiproteins:
    temp = Entrez.efetch( db='protein', 
                            id=ncbiproteins[protein], 
                            rettype="fasta", 
                            retmode="text")
    transgeneProteins[protein] = str( Bio.SeqIO.read( temp, 'fasta' ).seq )
    
# phi29 protein sequence from http://www.uniprot.org/uniprot/P03680
transgeneProteins[ 'phi29pol' ] = str( Bio.SeqIO.read( 'seq_and_annotation/phi29pol.fa', 'fasta' ).seq )
# yfp0 nucleotide sequence from Subramaniam et al. PNAS 2013
transgeneProteins[ 'yfp' ] = str( Bio.SeqIO.read( 'seq_and_annotation/yfp0.fasta', 'fasta' ).seq.translate()[:-1] )

for caiValue in numpy.arange(0.1,1,0.1): 
    seqrecords = list()
    for protein in transgeneProteins:
        #find the nucleotide sequence with cai value of 1.
        maxCaiVariant = list()
        for aa in transgeneProteins[protein]:
            for codon in cognateCodons[ Bio.SeqUtils.seq3(aa).upper() ]:
                if sharpEcoliIndex[ codon ] == 1:
                    break
            maxCaiVariant.append(codon)
        maxCaiVariant = ''.join(maxCaiVariant)
        
        #create 10 random variants with the required cai value.
        for variant in range(10):
            newSeq = getSeqForTargetCai( maxCaiVariant, caiValue )
            fasta_id = protein + "_cai_" + "%0.1f"%caiValue + "_variant_" + str(variant+1)            
            seqrecords.append( SeqRecord( Bio.Seq.Seq( newSeq), id=fasta_id, description='') )
            
    outputFile = "seq_and_annotation/Ecoli.transgene.cai" + "%0.1f"%caiValue + ".fasta"
    Bio.SeqIO.write( seqrecords, outputFile, "fasta" )
clear_output()
````

	// Initialize R_grid and avgRibosomeDensityGrid. 
	// R_grid now contains the id of ribosome at each mRNA position.
	// If there is no ribosome then that position get the value totalNumberOfRibosomes instead of 0 as 0 is a ribosome id
	// The state of the system with respect to mRNAs and bound ribosomes
	// avgRibosomeDensityGrid keeps track of the average ribosome density at each position of each mRNA after threshold time has elapsed.
	
	int** R_grid;							
	int** numberOfElongationsGrid;							
	float** avgRibosomeDensityGrid;							
	R_grid = malloc(sizeof(int *) * totalNumberOfMrna);
	numberOfElongationsGrid = malloc(sizeof(int *) * totalNumberOfMrna);
	avgRibosomeDensityGrid = (float**) malloc(sizeof(float *) * totalNumberOfMrna);
	for(c1=0;c1<totalNumberOfMrna;c1++)
		{	
		R_grid[c1] = malloc(sizeof(int *) * obs_max_len);
		numberOfElongationsGrid[c1] = malloc(sizeof(int *) * obs_max_len);
		avgRibosomeDensityGrid[c1] = (float*) malloc(sizeof(float *) * obs_max_len);
		for(c2=0;c2<Gene[mRNA[c1].gene].length;c2++)
			{	
			R_grid[c1][c2] = totalNumberOfRibosomes;
			avgRibosomeDensityGrid[c1][c2] = 0.0;
			numberOfElongationsGrid[c1][c2] = 0;
			}
		}
	
	// Initialize free mRNAs
	c3=0;
	for(c2=0;c2<totalNumberOfGenes;c2++)
	{	for(c1=0;c1<Gene[c2].exp;c1++)
		{	freeMrnaList[c2][c1]=c3;											// At the begininning all mRNAs can be initiated (store their ids here)
			c3++;
		}
	}
	
	// Initialize elongation time vectors and number of elongatable bound ribosomes to each codon.
	int codonloop;
	for(codonloop=0;codonloop<61;codonloop++)
	{	
		summedElongationTimes[codonloop] = 0.0;
		numberOfElongations[codonloop] = 0;
		numberOfAbortions[codonloop] = 0;
		n_Rb_e[codonloop] = 0;														
	}

	// Scale tRNA concentration to match the total number of tRNAs at that growth rate.
	// Scale tRNA aminoacylation rate and ribosome associate rate to change units.
	int trnaloop;
	double temp_trnatotal = 0;
	for(trnaloop=0;trnaloop<61;trnaloop++)
	{	
		temp_trnatotal += Trna[trnaloop].concn;
	}
	for(trnaloop=0;trnaloop<61;trnaloop++)
	{	
		Trna[trnaloop].concn *= (totalNumberOfTrna / temp_trnatotal);
		Trna[trnaloop].aminoacylationRateConstant = Trna[trnaloop].aminoacylationRateConstant / (cellVolume * 1000 * AVOGADRO_NUMBER); // converting from (moles per litre)-1s-1 to (molecules per cell)-1s-1 units.
		Trna[trnaloop].trnaRibosomeRateConstant = ribosomeTrnaKcatByKm / (cellVolume * 1000 * AVOGADRO_NUMBER); // converting from (moles per litre)-1s-1 to (molecules per cell)-1s-1 units.
	}

	// Initialize the pools of deacylated, acylated and ribosome bound tRNAs.
	// All trnas are initially free and deacylated.
	for(trnaloop=0;trnaloop<61;trnaloop++)
	{	
		deacylatedFreeTrna[trnaloop] = Trna[trnaloop].concn;
		acylatedFreeTrna[trnaloop] = 0;
		ribosomeBoundTrna[trnaloop] = 0;
		totalTrna[trnaloop] = deacylatedFreeTrna[trnaloop] + acylatedFreeTrna[trnaloop] + ribosomeBoundTrna[trnaloop];
		avgAcylatedTrnaConcn[trnaloop] = 0.0;
		avgDeacylatedTrnaConcn[trnaloop] = 0.0;
		avgRibosomeBoundTrnaConcn[trnaloop] = 0.0;
	}

	/////////////////////////////////////////////////
	// Begin the actual continuous time simulation
	/////////////////////////////////////////////////

	while(currentTime<totalTime)														// Till current time is less than max time
	{
		// Calculate the total rate of Gillespie simulation at this step as the sum of the rates of initiation, elongation, aminoacylation and abortion.
		totalRate = 0.0;

		// Total Aminoacylation Rate
		totalAminoacylationRate = 0.0;
		for(trnaloop=0;trnaloop<61;trnaloop++)
			{	
			totalAminoacylationRate += ( Trna[trnaloop].aminoacylationRateConstant * deacylatedFreeTrna[trnaloop] );
			}
		totalRate += totalAminoacylationRate;

		// Total Initiation Rate
		totalInitiationRate = totalInitiationScaledMrnaCopyNumber*Rf/( (1-boundRibosomeFraction) * totalNumberOfRibosomes );				
		totalRate += totalInitiationRate;

		// Total Elongation and Abortion Rates
		totalElongationRate = 0.0;
		for(codonloop=0;codonloop<61;codonloop++)
		{	
			codon_dependent_elongation_rate[codonloop] = 0.0;
			codon_elongation_rate[codonloop] = 0.0;
			abortionRate[codonloop] = 0.0;
			for (cognatetrnaloop=0; cognatetrnaloop<2; cognatetrnaloop++)
				{
				codon_dependent_elongation_rate[codonloop] += ( Trna[ Codon[codonloop].trna[cognatetrnaloop] ].trnaRibosomeRateConstant * acylatedFreeTrna[ Codon[codonloop].trna[cognatetrnaloop] ] * Codon[codonloop].weight[cognatetrnaloop] );
				}
			codon_elongation_rate[codonloop] = 1 / (codonMinElongationTime + 1 / codon_dependent_elongation_rate[codonloop] );

			if ( codon_elongation_rate[codonloop] < abortionThresholdRate  ) // Abortion rate is non-zero only if the codon_elongation_rate falls below the threshold for abortion. 
				{
				abortionRate[codonloop] = abortionRateConstant; 
				}

			r_elng[codonloop] = ( codon_elongation_rate[codonloop] + abortionRate[codonloop] ) * n_Rb_e[codonloop];
			totalElongationRate += r_elng[codonloop];
	
		}
		totalRate += totalElongationRate;



		// The totalTrna values below should change at all. This condition can be useful for debugging.
		temp_trna_total = 0;	
		for(trnaloop=0;trnaloop<61;trnaloop++)
		{	
			totalTrna[trnaloop] = deacylatedFreeTrna[trnaloop] + acylatedFreeTrna[trnaloop] + ribosomeBoundTrna[trnaloop];
			temp_trna_total += totalTrna[trnaloop];
		}
		temp_trna = temp_trna_total;

		// Increment time. Following Shah et al, we have not implemented the strict Gillespie prescription of random time drawn from an exponential distribution. Since all calculated quantities are steady-state values, this difference does not matter.
		//
		currentTime +=  ( 1 / totalRate );

		// Tracking time-averaged free ribosomes and tRNAs for final output.
		if(currentTime>thresholdTime)
		{	if(printOptions[3]==1)				
				{	
				for(trnaloop=0;trnaloop<61;trnaloop++)
					{	
					avgAcylatedTrnaConcn[trnaloop] += (double)acylatedFreeTrna[trnaloop] / totalRate;
					avgDeacylatedTrnaConcn[trnaloop] += (double)deacylatedFreeTrna[trnaloop] / totalRate;
					avgRibosomeBoundTrnaConcn[trnaloop] += (double)ribosomeBoundTrna[trnaloop] / totalRate;
					}
				avg_Rf += (double)Rf / totalRate;
				}
		}
		
		// Draw a random reaction channel by first choosing between aminoacylation, initiation or elongation and then picking the sub channel according to the respective rate.
			
		randomRate = totalRate * rand()/((double)RAND_MAX);	// Pick a random uniform to pick an event
		previousRate = 0.0;
		currentRate = 0.0;

		// Aminoacylation
		previousRate = currentRate;
		currentRate += totalAminoacylationRate;							
		// Pick tRNA that is aminoacylated with a probability proportional to its aminoacylation rate.	
		if(randomRate > previousRate && randomRate < currentRate)
			{
			for(trnaloop=0;trnaloop<61;trnaloop++)	
				{	
				previousRate += ( Trna[trnaloop].aminoacylationRateConstant * deacylatedFreeTrna[trnaloop] );
				if(randomRate < previousRate)
					{	
					deacylatedFreeTrna[trnaloop]--; // Decrease deacylated trna by one.
					acylatedFreeTrna[trnaloop]++; // Increase acylated trna by one.
					break;
					}
				}
			}

		// Translation initiation
		previousRate = currentRate;
		currentRate += totalInitiationRate;							
	
		if(randomRate > previousRate && randomRate < currentRate)
		{	
			// Ribosomes are picked sequentially
			r_id = next_avail_ribo;
			Ribosome[r_id].pos = 0;
			Ribosome[r_id].timeOfLastInitiation = currentTime;
			Ribosome[r_id].timeOfLastElongation = currentTime;
			Ribosome[r_id].timeOfElongatableSwitch = currentTime;
			next_avail_ribo++;
			
			// Pick the gene for mRNA based on their initiation rates.
			for(geneloop=0;geneloop<totalNumberOfGenes;geneloop++)
				{	
				previousRate += initiationScaledMrnaCopyNumber[geneloop] *Rf / ( (1-boundRibosomeFraction) * totalNumberOfRibosomes );				

				if(randomRate < previousRate)
					{	
					c2 = rand()%freeMrna[geneloop];										// Once a gene is selected pick a random mRNA
					m_id = freeMrnaList[geneloop][c2];
					freeMrna[geneloop]--;
					totalFreeMrna--;
					initiationScaledMrnaCopyNumber[geneloop]-=Gene[geneloop].ini_prob;
					totalInitiationScaledMrnaCopyNumber-=Gene[geneloop].ini_prob;

					if(c2!=freeMrna[geneloop])
						{	
						freeMrnaList[geneloop][c2] = freeMrnaList[geneloop][freeMrna[geneloop]];			// Accounting of free mRNAs for next round
						}
					break;
					}
				}			
			
			if(currentTime>thresholdTime)
			{	mRNA[m_id].numberOfInitiations++;											// Store the # of initn events on this mRNA
				mRNA[m_id].averageInitiationTime += currentTime-mRNA[m_id].timeOfLastMrnaInitiation;		// Time between initn
			}	
			mRNA[m_id].timeOfLastMrnaInitiation = currentTime;

			c_id = Gene[mRNA[m_id].gene].seq[0];							// Codon at pos 1
	
			Ribosome[r_id].mRNA = m_id;							

			if(R_grid[m_id][10]==totalNumberOfRibosomes)									// Check if the current ribosome can be elongated
			{	Rb_e[c_id][n_Rb_e[c_id]] = r_id;							// If no ribosome at pos+10 then it can
				Ribosome[r_id].elng_cod_list = c_id;
				Ribosome[r_id].elng_pos_list = n_Rb_e[c_id];
				n_Rb_e[c_id]++;
			}
			
			R_grid[m_id][0] = r_id;												// Update the ribosome grid upon initiation
			
			Rf--;																// Update number of free ribosomes
		}
		
		// Translation Elongation
		previousRate = currentRate;
		currentRate += totalElongationRate;							
	
		if(randomRate > previousRate && randomRate < currentRate)
		{
			// Pick a random codon based on respective ( elongation + abortion rate )	
			for(codonloop=0;codonloop<61;codonloop++)	
				{	
				previousRate += r_elng[codonloop];
				if(randomRate < previousRate)
					{	
					c_id = codonloop;
					break;
					}
				}
			
			// Pick between elongation and abortion based on the respective probability if abortion rate is non-zero.
			abort_flag = 0;
			if (abortionRate[c_id] > 0)
				{
				randomAbortionRate = (codon_elongation_rate[c_id] + abortionRate[c_id] ) * rand()/((double)RAND_MAX); 
				if ( randomAbortionRate < abortionRate[c_id] )
					{
					abort_flag = 1;
					}
				}
    
			x = rand()%n_Rb_e[c_id];											// Randomly pick an elongatable ribosome bound to codon c_id
			r_id = Rb_e[c_id][x];
			m_id = Ribosome[r_id].mRNA;
		
			if(Ribosome[r_id].pos>0) // If the ribosome is at the 2nd codon or higher (Ribosomes at 1st codon have empty P-sites).
			{	
				deacylatedFreeTrna[Ribosome[r_id].tRNA]++; // Move the tRNA at the P-site into the E-site and release it into deacylated free pool.
				ribosomeBoundTrna[Ribosome[r_id].tRNA]--; 
			}

			if (abort_flag == 0) // If aborting, no need to pick a tRNA from the acylated pool.
				{
	            // Pick a random A-site cognate tRNA based on respective elongation probabilities.
	            coin = rand()/((double)RAND_MAX);
	            prob_g = 0.0;
	    
	            for(cognatetrnaloop=0;cognatetrnaloop<2;cognatetrnaloop++)
	            {   
	                prob_g += ( Trna[ Codon[c_id].trna[cognatetrnaloop] ].trnaRibosomeRateConstant * acylatedFreeTrna[ Codon[c_id].trna[cognatetrnaloop] ] * Codon[c_id].weight[cognatetrnaloop] );
	    
	                if(coin<prob_g)
	                {   
	                    cognate_trna_id = Codon[c_id].trna[cognatetrnaloop]; 
	                    break;
	                }   
	            }  	
		
				Ribosome[r_id].tRNA = cognate_trna_id; // Set the tRNA on the ribosome to be the chosen cognate tRNA.
				acylatedFreeTrna[Ribosome[r_id].tRNA]--; // Account for the tRNA that entered the ribosome from the free pool.
				ribosomeBoundTrna[Ribosome[r_id].tRNA]++; // Account for the tRNA that entered the ribosome from the free pool.
				}

			if(currentTime > thresholdTime)
				{	
				if(printOptions[0]==1)
					{	
					summedElongationTimes[c_id] += currentTime-Ribosome[r_id].timeOfElongatableSwitch;				// For estimation of avg elongation times of codons
					numberOfElongations[c_id]++;
					if (abort_flag == 1)
						{
							numberOfAbortions[c_id]++;
						}	
					}
				if(printOptions[4]==1)
					{	
					avgRibosomeDensityGrid[m_id][Ribosome[r_id].pos] += currentTime-Ribosome[r_id].timeOfLastElongation;		// Keep track of the ribosome density at the corresponding codon position. 
					numberOfElongationsGrid[m_id][Ribosome[r_id].pos] ++;		// Keep track of the number of times that this position on this mRNA was elongated. 
					}
				}

			if ( Ribosome[r_id].pos == (Gene[mRNA[m_id].gene].length-1) || (abort_flag == 1 ) )					// Check if the current elongation has led to termination or abortion of translation.
				{	
				R_grid[m_id][Ribosome[r_id].pos] = totalNumberOfRibosomes;						// Update the position of ribosomes on the mRNA.
				if (abort_flag == 0) // If aborting, no trna was incorporated into the ribosome above.
					{
					deacylatedFreeTrna[Ribosome[r_id].tRNA]++; // Release the ribosome bound tRNA back into the deacylated tRNA pool.
					ribosomeBoundTrna[Ribosome[r_id].tRNA]--; // Release the ribosome bound tRNA back into the deacylated tRNA pool.
					}

				Ribosome[r_id].tRNA = 0; // Reset the tRNA index on the ribosome. Note that this is swapped with the next_avail_ribosome value below.
				Ribosome[r_id].pos++;
				Rf++;															// Free a ribosome upon termination

				n_Rb_e[c_id]--;
				if(x!=n_Rb_e[c_id])
				{	Ribosome[Rb_e[c_id][n_Rb_e[c_id]]].elng_pos_list = x;			// Update the ids and number of elongatable ribosomes
					Rb_e[c_id][x] = Rb_e[c_id][n_Rb_e[c_id]];
				}
			
				if(currentTime>thresholdTime)
					{	
					if (abort_flag == 1)
						{
						mRNA[m_id].numberOfAbortedProteins++;
						mRNA[m_id].averageAbortionTime += currentTime-Ribosome[r_id].timeOfLastInitiation;	// Update the time to abortion
						}
					else if (abort_flag == 0)
						{
						mRNA[m_id].numberOfFullProteins++;										// Update the number of trans evnts on curr mRNA
						mRNA[m_id].averageElongationTime += currentTime-Ribosome[r_id].timeOfLastInitiation;	// Update the time to translation
						}
				}

				// Update any previously unelongatable ribosomes 
				if(Ribosome[r_id].pos > 10 && R_grid[m_id][Ribosome[r_id].pos-11]<totalNumberOfRibosomes)						// When the ribosome moves, a previously unelongatable ribosome
					{
					c2_id = Gene[mRNA[m_id].gene].seq[Ribosome[r_id].pos-11];			// can now be elongatable on the same mRNA if its 11 codon behind
					Rb_e[c2_id][n_Rb_e[c2_id]] = R_grid[m_id][Ribosome[r_id].pos-11];
					
					Ribosome[R_grid[m_id][Ribosome[r_id].pos-11]].timeOfElongatableSwitch = currentTime;
					Ribosome[R_grid[m_id][Ribosome[r_id].pos-11]].elng_cod_list = c2_id;
					Ribosome[R_grid[m_id][Ribosome[r_id].pos-11]].elng_pos_list = n_Rb_e[c2_id];
					n_Rb_e[c2_id]++;

					}
								
				if ( Ribosome[r_id].pos <= 10 )													// If the current position is less than or equal to codon 11, make the current mRNA initiable
					{
					g_id = mRNA[m_id].gene;
					freeMrnaList[g_id][freeMrna[g_id]] = m_id;
					freeMrna[g_id]++;
					totalFreeMrna++;
					initiationScaledMrnaCopyNumber[g_id] += Gene[g_id].ini_prob;
					totalInitiationScaledMrnaCopyNumber += Gene[g_id].ini_prob;
					}

				// Update the ribosomes avail for initiation
				next_avail_ribo--;
				
				if(r_id!=next_avail_ribo)											// The last initiated ribosome's data is swapped with
					{	
					Ribosome[r_id].mRNA = Ribosome[next_avail_ribo].mRNA;					// the ribosome that just finished translation (terminated)
					Ribosome[r_id].pos = Ribosome[next_avail_ribo].pos;						// This is done to primarily keep track of ONLY ribosomes

					Ribosome[r_id].tRNA = Ribosome[next_avail_ribo].tRNA;	

					Ribosome[r_id].timeOfLastInitiation = Ribosome[next_avail_ribo].timeOfLastInitiation;		// that are currently bound for faster computation

					Ribosome[r_id].timeOfLastElongation = Ribosome[next_avail_ribo].timeOfLastElongation;
					Ribosome[r_id].timeOfElongatableSwitch = Ribosome[next_avail_ribo].timeOfElongatableSwitch;
				
					Ribosome[r_id].elng_cod_list = Gene[mRNA[Ribosome[next_avail_ribo].mRNA].gene].seq[Ribosome[next_avail_ribo].pos];
					Ribosome[r_id].elng_pos_list = Ribosome[next_avail_ribo].elng_pos_list;
					R_grid[Ribosome[r_id].mRNA][Ribosome[r_id].pos] = r_id;
				
					if(R_grid[Ribosome[r_id].mRNA][Ribosome[r_id].pos+10]==totalNumberOfRibosomes || (Ribosome[r_id].pos+10)>=Gene[mRNA[Ribosome[r_id].mRNA].gene].length)
						{
						Rb_e[Ribosome[r_id].elng_cod_list][Ribosome[r_id].elng_pos_list] = r_id;
						}
					}
				termtn_now = 1;
				}

			else if((Ribosome[r_id].pos+11)>=Gene[mRNA[m_id].gene].length || R_grid[m_id][Ribosome[r_id].pos+11]==totalNumberOfRibosomes)	// Check if the ribosome is still elongatable
			{	R_grid[m_id][Ribosome[r_id].pos] = totalNumberOfRibosomes;															// Update the position of ribosomes on the mRNA
				
				Ribosome[r_id].pos++;
				R_grid[m_id][Ribosome[r_id].pos] = r_id;
				c2_id = Gene[mRNA[m_id].gene].seq[Ribosome[r_id].pos];
					
				if(c2_id!=c_id)															// If the codon has changed shift the elongatable ribosome
				{	Rb_e[c2_id][n_Rb_e[c2_id]] = r_id;									// to the other codon
					Ribosome[r_id].elng_cod_list = c2_id;
					Ribosome[r_id].elng_pos_list = n_Rb_e[c2_id];
					n_Rb_e[c2_id]++;
					
					n_Rb_e[c_id]--;
					if(x!=n_Rb_e[c_id])
					{	Ribosome[Rb_e[c_id][n_Rb_e[c_id]]].elng_pos_list = x;				// Update the ids and number of elongatable ribosomes
						Rb_e[c_id][x] = Rb_e[c_id][n_Rb_e[c_id]];
					}
				}
			}
			
			else																		// When ribosome is not elongatable anymore
			{	R_grid[m_id][Ribosome[r_id].pos] = totalNumberOfRibosomes;								// Update the position of ribosomes on the mRNA
				
				Ribosome[r_id].pos++;
				R_grid[m_id][Ribosome[r_id].pos] = r_id;
				Ribosome[r_id].elng_cod_list = Gene[mRNA[m_id].gene].seq[Ribosome[r_id].pos];

				n_Rb_e[c_id]--;
				if(x!=n_Rb_e[c_id])
				{	Ribosome[Rb_e[c_id][n_Rb_e[c_id]]].elng_pos_list = x;					// Update the ids and number of elongatable ribosomes
					Rb_e[c_id][x] = Rb_e[c_id][n_Rb_e[c_id]];
				}
				
			}
			
			if(termtn_now==0)															// When elongation does not lead to termination
			{	if(Ribosome[r_id].pos>10 && R_grid[m_id][Ribosome[r_id].pos-11]<totalNumberOfRibosomes)		// When the ribosome moves, a previously unelongatable ribosome
				{	c2_id = Gene[mRNA[m_id].gene].seq[Ribosome[r_id].pos-11];				// can now be elongatable on the same mRNA if its 11 codon behind
					Rb_e[c2_id][n_Rb_e[c2_id]] = R_grid[m_id][Ribosome[r_id].pos-11];
					
					Ribosome[R_grid[m_id][Ribosome[r_id].pos-11]].timeOfElongatableSwitch = currentTime;
					Ribosome[R_grid[m_id][Ribosome[r_id].pos-11]].elng_cod_list = c2_id;
					Ribosome[R_grid[m_id][Ribosome[r_id].pos-11]].elng_pos_list = n_Rb_e[c2_id];
					n_Rb_e[c2_id]++;
					
				}
				if(Ribosome[r_id].pos==10)													// If the current position is at codon 11 make the current mRNA initiable
				{	g_id = mRNA[m_id].gene;
					freeMrnaList[g_id][freeMrna[g_id]] = m_id;
					freeMrna[g_id]++;
					totalFreeMrna++;
					initiationScaledMrnaCopyNumber[g_id] += Gene[g_id].ini_prob;
					totalInitiationScaledMrnaCopyNumber += Gene[g_id].ini_prob;
				}

				Ribosome[r_id].timeOfLastElongation = currentTime;												// Upon elongation, update the elong ini time for the next evnt
				Ribosome[r_id].timeOfElongatableSwitch = currentTime;												// Upon elongation, update the elong ini time for the next evnt
			}
			else
			{	termtn_now = 0;
			}
			
			if(currentTime>((double)t_print) && printOptions[6]==1)
			{	c3 = 0;
				strcpy(outfile,outfilePrefix);
				sprintf(tmp_t,"%d",t_print);
				f1 = fopen(strcat(strcat(outfile,"_ribo_pos_"),tmp_t),"w");			// Print ribosome pos every second
				printf("%s",outfile);	
				for(c1=0;c1<totalNumberOfGenes;c1++)
				{	for(c4=0;c4<Gene[c1].exp;c4++)
					{	for(c2=0;c2<Gene[c1].length;c2++)
						{	if(R_grid[c3][c2]==totalNumberOfRibosomes)
							{	fprintf(f1,"0 ");
							}
							else
							{	fprintf(f1,"1 ");
							}
						}
						c3++;
						fprintf(f1,"\n");
					}
				}
				fclose(f1);
				t_print++;
			}
		}
	if (number_of_iterations%10000000 == 0)
		{

				{	
				n_trans[c1] += mRNA[c3].numberOfFullProteins;
				n_aborts[c1] += mRNA[c3].numberOfAbortedProteins;
				g_ini[c1] += mRNA[c3].averageInitiationTime;

				g_atimes[c1] += mRNA[c3].averageAbortionTime;

				c3++;
				}
		}
	}

	
	// Print the output

	// Elongation times of all codons
	if(printOptions[0]==1)

	if(printOptions[3]==1)
		{

		fprintf(f5,"Ribosome\t%0.6f\t%0.6f\n",avg_Rf,1-avg_Rf);
		double acylatedFraction;

			{

				avgRibosomeBoundTrnaConcn[trnaloop] = avgRibosomeBoundTrnaConcn[trnaloop]/(totalTime-thresholdTime)/totalTrna[trnaloop];

				}
			}

		}
	
	// Time-averaged ribosome density on each gene that was tracked after threshold time was crossed.
	if(printOptions[4]==1)
		{	
		strcpy(outfile,outfilePrefix);
		f6 = fopen(strcat(outfile,"_gene_ribo_density.out"),"w");
		strcpy(outfile,outfilePrefix);
		f6a = fopen(strcat(outfile,"_gene_number_elongs.out"),"w");
		
		float** geneSteadyStateRibosomeDensity;	// Sum up the ribosome density on all mRNAs for each gene. 
		int** geneTotalNumberOfElongations;	// Sum up the total number of elongations on all mRNAs for each gene.
		geneSteadyStateRibosomeDensity = (float**) malloc(sizeof(float *) * totalNumberOfGenes);
		geneTotalNumberOfElongations = (int**) malloc(sizeof(int *) * totalNumberOfGenes);

		m_id = 0;

			geneTotalNumberOfElongations[geneloop] = (int*) malloc(sizeof(int *) * obs_max_len);
			start_mid = m_id;
			for(c1=0; c1<Gene[geneloop].length; c1++)

				geneSteadyStateRibosomeDensity[geneloop][c1] = 0.0;
				geneTotalNumberOfElongations[geneloop][c1] = 0;
				for (c2=0; c2<Gene[geneloop].exp; c2++)
					{	
					geneSteadyStateRibosomeDensity[geneloop][c1] += avgRibosomeDensityGrid[start_mid+c2][c1];
					geneTotalNumberOfElongations[geneloop][c1] += numberOfElongationsGrid[start_mid+c2][c1];
					if (c1 == 0)
						{
						m_id++;
						}
					}
				}
			}

		for(geneloop=0; geneloop<totalNumberOfGenes; geneloop++)
			{	
			for(c2=0;c2<Gene[geneloop].length;c2++)
				{	
				fprintf(f6,"%0.2f ", geneSteadyStateRibosomeDensity[geneloop][c2]);
				fprintf(f6a,"%d ", geneTotalNumberOfElongations[geneloop][c2]);
				}
			fprintf(f6,"\n");
			fprintf(f6a,"\n");
			}

		free( geneTotalNumberOfElongations );

    time_t result = time(NULL);

				}

	fprintf(fh, "trna_id\tanticodon\tconcn\tribosome_association_rate\taminoacylation_rate\n" );

		if (Trna[c1].concn > 0)
			{
			fprintf(fh, "%d\t%s\t%0.0f\t%0.3g\t%0.3g\n", c1, Trna[c1].anticodon,

					Codon[c1].trna[0], Codon[c1].trna[1], Codon[c1].weight[0], Codon[c1].weight[1] );

    fh = fopen( strcat( outfile, "_global_vars_data.in" ), "w" );

````python
%%writefile scripts/run_simulation.py

import sys
import subprocess
import numpy

jobindex = 0

# Simulations with differential aminoacylation rates for leucine and serine trna and nonzero time for intraribosomal events.
# (Figures 3,5,7)
for (aminoacid, relativetrnaaminoacylationrate) in zip( ['leu','ser'], ['0.5 1.5 0.5 0.5 1', '0.5 1.5 1 0.5'] ):
    for (ribotrnakcatkm, codonminelongtime) in zip( ["2e7"], ["0.05"] ):
        for ribosomes in [44000]:
            for starvationlevel in [0.01] + list( (2.0)**(numpy.arange(0,-10,-0.5))):
                for abortionrate in [0.00, 0.1] + list((2.0)**(numpy.arange(5.0,-10.0,-1.0))):
                    jobindex += 1
                    if not jobindex == int(sys.argv[1]):
                        continue
                    subprocess.call( [ "./scripts/ecoli_translation" + 
                                      " --threshold-time 1000" + 
                                      " --total-time 1500" +
                                      " --aminoacid " + aminoacid  + " starvationfactor " + "%0.6f"%starvationlevel +
                                      " --total-ribosomes " + str(ribosomes) + 
                                      " --aminoacid " + aminoacid + " relativetrnaaminoacylationrates " + relativetrnaaminoacylationrate + 
                                      " --abort-threshold-rate 10" + 
                                      " --abortion-relative-rate " + "%0.6f"%abortionrate +
                                      " --ribotrnakcatkm " + ribotrnakcatkm +
                                      " --min-elong-time " + codonminelongtime + 
                                      " --output-prefix output_files/simulation_v1/output_v1" +  
                                      " --input-genes seq_and_annotation/Ecoli.input.seq"  ],
                                    shell= True )
                    
print jobindex                    
                    
# Simulations to compare nonzero and zero time for intraribosomal events (Figure 3)
for (ribotrnakcatkm, codonminelongtime) in zip( ["2e7","2e6"], ["0.05", "0.00"] ):
    for ribosomes in [44000]:
        for abortionrate in [0.1]:
            jobindex += 1
            if not jobindex == int(sys.argv[1]):
                continue
            subprocess.call( [ "./scripts/ecoli_translation" + 
                              " --threshold-time 1000" + 
                              " --total-time 1500" +
                              " --total-ribosomes " + str(ribosomes) + 
                              " --abort-threshold-rate 10" + 
                              " --abortion-relative-rate " + "%0.6f"%abortionrate +
                              " --ribotrnakcatkm " + ribotrnakcatkm +
                              " --min-elong-time " + codonminelongtime + 
                              " --output-prefix output_files/simulation_v2/output_v2" +  
                              " --input-genes seq_and_annotation/Ecoli.input.seq"  ],
                            shell= True )
print jobindex                    

# Simulations to overexpress transgenes (Figure 3)
for (ribotrnakcatkm, codonminelongtime) in zip( ["2e7","2e6"], ["0.05", "0.00"] ):
    for ribosomes in [44000]:
        for abortionrate in [0.1]:
            for mrnaFraction in map( str, [1, 10, 50] ):
                for cai in map( lambda x: "%0.1f"%x, numpy.arange(0.1, 1, 0.1) ):
                    jobindex += 1
                    if not jobindex == int(sys.argv[1]):
                        continue
                    subprocess.call( [ "./scripts/ecoli_translation" + 
                                      " --threshold-time 1000" + 
                                      " --total-time 1500" +
                                      " --total-ribosomes " + str(ribosomes) + 
                                      " --abort-threshold-rate 10" + 
                                      " --abortion-relative-rate " + "%0.6f"%abortionrate +
                                      " --ribotrnakcatkm " + ribotrnakcatkm +
                                      " --min-elong-time " + codonminelongtime + 
                                      " --output-prefix output_files/simulation_v3/output_v3_mrnafraction_" + mrnaFraction + "_cai_" + cai + 
                                      " --input-genes seq_and_annotation/Ecoli.input.transgenefrac" + mrnaFraction + ".cai" + cai + ".seq"   ],
                                    shell= True )
print jobindex         

# Simulations with equal aminoacylation rates for leucine and serine trna. (Figure 3)
for (aminoacid, relativetrnaaminoacylationrate) in zip( ['leu','ser'], ['1 1 1 1 1','1 1 1 1'] ):
    for (ribotrnakcatkm, codonminelongtime) in zip( ["2e7"], ["0.05"] ):
        for ribosomes in [44000]:
            for starvationlevel in list( (2.0)**(numpy.arange(0,-10,-0.5))):                
                for abortionrate in [0.1]:
                    jobindex += 1
                    if not jobindex == int(sys.argv[1]):
                        continue
                    subprocess.call( [ "./scripts/ecoli_translation" + 
                                      " --threshold-time 1000" + 
                                      " --total-time 1500" +
                                      " --aminoacid " + aminoacid  + " starvationfactor " + "%0.6f"%starvationlevel +
                                      " --total-ribosomes " + str(ribosomes) + 
                                      " --aminoacid " + aminoacid + " relativetrnaaminoacylationrates " + relativetrnaaminoacylationrate + 
                                      " --abort-threshold-rate 10" + 
                                      " --abortion-relative-rate " + "%0.6f"%abortionrate +
                                      " --ribotrnakcatkm " + ribotrnakcatkm +
                                      " --min-elong-time " + codonminelongtime + 
                                      " --output-prefix output_files/simulation_v4/output_v4" +  
                                      " --input-genes seq_and_annotation/Ecoli.input.seq"  ],
                                    shell= True )
print jobindex                    

# Simulations of leucine starvation with codon-trna kcat/KM rates from Sorensen et al. J. Mol. Biol. 2005.
# (Figure S2C). Copy ecoli_translation.c to ecoli_translation_sorensen.c with the input cognate weights file
# changed in the code. Compile this as ecoli_translation_sorensen and then run the loop below.
for (aminoacid, relativetrnaaminoacylationrate) in zip( ['leu'], ['1 1 1 1 1'] ):
    for (ribotrnakcatkm, codonminelongtime) in zip( ["2e7"], ["0.05"] ):
        for ribosomes in [44000]:
            for starvationlevel in [0.01] + list( (2.0)**(numpy.arange(0,-10,-0.5))):
                for abortionrate in [0.1]:
                    jobindex += 1
                    if not jobindex == int(sys.argv[1]):
                        continue
                    subprocess.call( [ "./scripts/ecoli_translation_sorensen" + 
                                      " --threshold-time 1000" + 
                                      " --total-time 1500" +
                                      " --aminoacid " + aminoacid  + " starvationfactor " + "%0.6f"%starvationlevel +
                                      " --total-ribosomes " + str(ribosomes) + 
                                      " --aminoacid " + aminoacid + " relativetrnaaminoacylationrates " + relativetrnaaminoacylationrate + 
                                      " --abort-threshold-rate 10" + 
                                      " --abortion-relative-rate " + "%0.6f"%abortionrate +
                                      " --ribotrnakcatkm " + ribotrnakcatkm +
                                      " --min-elong-time " + codonminelongtime + 
                                      " --output-prefix output_files/simulation_v5/output_v5" +  
                                      " --input-genes seq_and_annotation/Ecoli.input.seq"  ],
                                    shell= True )
print jobindex  

#Simulation to vary the difference in aminoacylation rates between different tRNAs (Figure S2D)
p = numpy.arange(0,1,0.1)
relativeaminoacylationrates = []
for row in numpy.round(zip(1-p,1+p,1-p,1-p,numpy.ones(10,)),2):
    temp = ''
    for col in row:
        temp = temp + '%0.1f '%col
    relativeaminoacylationrates.append( temp.strip() )

for aminoacid in ['leu']:
    for relativetrnaaminoacylationrate in relativeaminoacylationrates:
        for (ribotrnakcatkm, codonminelongtime) in zip( ["2e7"], ["0.05"] ):
            for ribosomes in [44000]:
                for starvationlevel in [(2.0)**(-10)]:                
                    for abortionrate in [0.1]:
                        jobindex += 1
                        if not jobindex == int(sys.argv[1]):
                            continue
                        subprocess.call( [ "./scripts/ecoli_translation" + 
                                          " --threshold-time 1000" + 
                                          " --total-time 1500" +
                                          " --aminoacid " + aminoacid  + " starvationfactor " + "%0.6f"%starvationlevel +
                                          " --total-ribosomes " + str(ribosomes) + 
                                          " --aminoacid " + aminoacid + " relativetrnaaminoacylationrates " + relativetrnaaminoacylationrate + 
                                          " --abort-threshold-rate 10" + 
                                          " --abortion-relative-rate " + "%0.6f"%abortionrate +
                                          " --ribotrnakcatkm " + ribotrnakcatkm +
                                          " --min-elong-time " + codonminelongtime + 
                                          " --output-prefix output_files/simulation_v6/output_v6" +  
                                          " --input-genes seq_and_annotation/Ecoli.input.seq"  ],
                                        shell= True )
print jobindex 
````
