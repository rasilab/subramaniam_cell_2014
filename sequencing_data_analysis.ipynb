#### polyatrimmer.py ####

This script trims the polyA tails that are added during the library preparation step.

```python
%%writefile scripts/polyatrimmer.py

#!/usr/bin/env python

import re #regular expression library.
import argparse #to parse arguments.
import gzip #library to handle gzipped files.
from Bio.SeqIO.QualityIO import FastqGeneralIterator #for iterating through fasta records.

#parse arguments using the argparse module.
#----------------------
parser = argparse.ArgumentParser( description='Trim polyA runs and subsequent 3\' region from fastq single-end Illumina reads.\
                                    Required python modules: gzip, re, argparse, Bio.SeqIO.QualityIO. Output: trimmed reads fastq file and file containing reads that fall outside size cutoff.\
                                    Note that reads that do not have the minALength number of polyAs are discarded.' )
parser.add_argument( 'inputFile', metavar='INPUTFILE', help='fastq input filename, can be compressed in .gz format' )
parser.add_argument( '--minALength', dest='numberOfAs', type=int, default=10, help='minimum number of continuous As that triggers trimming (default: 10)' )
parser.add_argument( '--extraTrim', dest='extraTrimmingLength', type=int, default=1, help='trim these many nt 5\' to the polyA run (default: 1)' )
parser.add_argument( '--end5Trim', dest='end5TrimmingLength', type=int, default=1, help='trim these many nt at the 5\' end before writing output. Note that this length is not counted for the thresholds (default: 1) ' )
parser.add_argument( '--minResidualLength', dest='remainingLength', type=int, default=13, help='only retain reads that are at least this many nt long after discarding polyA runs + extraTrimmingLength (default: 13)' )

inputFile = parser.parse_args( ).inputFile
numberOfAs = parser.parse_args( ).numberOfAs
extraTrimmingLength = parser.parse_args( ).extraTrimmingLength
end5TrimmingLength = parser.parse_args( ).end5TrimmingLength
remainingLength = parser.parse_args( ).remainingLength
#----------------------

if inputFile.endswith('gz'): #optional parsing of gzip compressed files.
    recordIterator = FastqGeneralIterator( gzip.open( inputFile, 'r' ) )
    outputFile = open( inputFile.replace('fastq.gz','trimmed'), 'w' )
else:
    outputFile = open( inputFile.replace('fastq','trimmed'), 'w' )
    recordIterator = FastqGeneralIterator( open( inputFile, 'r' ) )

numberOfProcessedReads = 0
numberOfCountedReads = 0

for title, seq, qual in  recordIterator: #look at a fastq file for example of these lines.
    numberOfProcessedReads += 1
    stopLocation = re.search( 'A{' + str(numberOfAs) + ',}', seq ) #identifies reads with a run of >=numberOfAs adenines.
    if stopLocation is not None: #discard reads that do not have >=numberOfAs adenine stretch.
        trimStop = stopLocation.start() - extraTrimmingLength - 1 # trims the 'extraTrimmingLength' number of nt 5' to the polyA tail.
        if trimStop >= remainingLength: # ignores trimmed reads that are less than 'remainingLength' nt long.
            outputFile.write("@%s\n%s\n+\n%s\n" % (title, seq[end5TrimmingLength:trimStop + 1], qual[end5TrimmingLength:trimStop + 1])) # write the 5' end trimmed read to the output file.
            numberOfCountedReads += 1

outputFile.close()

print( '%d : Total number of processed reads'%numberOfProcessedReads)
print( '%2.2f : Percent of reads that passed threshold' % ( numberOfCountedReads / float( numberOfProcessedReads ) * 100 ))
```

#### collapsereads.py ####

This script finds the number of sequencing reads aligned by bowtie to each genomic location.

```python
%%writefile scripts/collapsereads.py

maxNumberOfMismatches = 2 # If there are more than this many mismatches, the read is not considered.
numberOfBasesToTrim = 8 # These many bases are trimmed from either side of the read before assigning ribosome occupancy.

import csv
import sys
import os
import cPickle as pickle
import time
import re

currentFile = sys.argv[1]
startTime = time.time()
countDict = dict()
readCount = 0
openedFile = open( currentFile )

reader = csv.DictReader( openedFile, delimiter='\t',
                    fieldnames=['readName', 'strand', 'ref', 'pos', 'seq',
                                'qual', 'Mparam', 'mismatchString'] )

for read in reader:

    readCount += 1
    readLength = len( read['seq'] )
    numberOfMismatches = read['mismatchString'].count(':')
    
    if ( numberOfMismatches<=maxNumberOfMismatches and readLength>2*numberOfBasesToTrim ): #length and mismatch cutoff.
        readStart = int( read['pos'] ) + numberOfBasesToTrim
        readEnd = int( read['pos'] ) + readLength - numberOfBasesToTrim
        readWeight = 1.0 / (readLength - 2 * numberOfBasesToTrim)

        if read['ref']+read['strand'] not in countDict:
            countDict[read['ref']+read['strand']] = dict()
        for readPos in range(readStart, readEnd):
            if readPos not in countDict[read['ref']+read['strand']]:
                countDict[read['ref']+read['strand']][readPos] = 0.0
            countDict[read['ref']+read['strand']][readPos] += readWeight

print readCount, ' reads collapsed.'

openedFile.close()
pickle.dump( countDict, open( currentFile.replace('.aligned','.pickled'), 'wb' ) )
```

#### calculate_genestats_and_samplestats.py ####

This calculates the ribosome occupancy around each codon and along mRNAs

```python
%%writefile scripts/calculate_genestats_and_samplestats.py

from Bio import SeqIO # For reading fasta files.
from Bio.SeqUtils import CodonUsage # A dictionary of codon names.
import sys # For displaying system calls.
import pandas # For tabular data processing.
import re # For regular expression processing.
import cPickle as pickle # For storing python data structures.
import numpy # For processing matrices and arrays.
import os # standard os commands.

dataFile = sys.argv[1]
offset = 150 # Number of nucleotides on either side of each codon where read density is analyzed.
thresholdReadDensity = 1.0 / 3.0 # Calculate codon statistics only if the read density for the gene exceeds this value.

genome = SeqIO.read( 'seq_and_annotation/NC_000913.fna', 'fasta' )
geneTable = pandas.read_table( 'seq_and_annotation/NC_000913.ptt', skiprows=2 )
geneTable['Start'] = geneTable['Location'].apply( lambda x: numpy.int32( x.split('.')[0] ) ) # Parse start coordinate of each CDS.
geneTable['End'] = geneTable['Location'].apply( lambda x: numpy.int32( x.split('.')[-1] ) ) #Part end coordinate of each CDS.

sampleName = dataFile.split('/')[-1].split( '.' )[0].lower()

print 'sample: ' + sampleName; sys.stdout.flush()

genomeCounts = dict()
dictData = pickle.load( open( dataFile ) )

for seq in dictData:
    genomeCounts[ seq[-1] ] = numpy.zeros( ( len(genome),) ) #-1 accounts for the fact that the + and - strands have separate keys in dictData.
    for pos in dictData[ seq ]:
        genomeCounts[ seq[-1] ][ pos ] = dictData[ seq ][ pos ]

genes = dict() # To keep track of calculated parameters at the single gene level.    
conditions = dict() # Boolean dictionary to implement all conditions.
rowCounts = 0 # To keep track of how many genes have been processed in each sample while code is running.

features = dict() # To keep track of calculated parameters at the single feature level. (Typically the feature is a codon).
features = CodonUsage.CodonsDict.copy()
features['start'] = dict()

profileDensity = numpy.zeros( (900,), dtype=float )
profileCounts = numpy.zeros( (900,), dtype=int)
frontSensitiveProfileDensity = numpy.zeros( (900,), dtype=float )
frontSensitiveProfileCounts = numpy.zeros( (900,), dtype=int)
midSensitiveProfileDensity = numpy.zeros( (900,), dtype=float )
midSensitiveProfileCounts = numpy.zeros( (900,), dtype=int)
robustProfileDensity = numpy.zeros( (900,), dtype=float )
robustProfileCounts = numpy.zeros( (900,), dtype=int)

for feature in features:
    features[ feature ] = dict()
    features[ feature ][ 'featureCounts' ] = 0 # This keeps track of the counts after thresholding genes.
    features[ feature ][ 'featureUsage' ] = 0.0 # This counts the usage of the feature in the total mRNA / monosome pool.
    # Add the read density and its square to the running sum of these values.
    features[ feature ][ 'readDensitySum' ] = numpy.zeros( (2*offset+1,) )
    features[ feature ][ 'readDensitySquareSum' ] = numpy.zeros( (2*offset+1,) )

for row in geneTable.iterrows():
    gene = row[1]['Gene']
    if gene in genes:
        gene = gene + row[1]['Synonym']
    if re.search( 'prfB|nrdF', gene ): # The length of these gene sequences is not a multiple of 3 and are excluded from analysis
        continue

    start = row[1]['Start'] - 1 #1-based .ptt file indexing to 0-based python indexing.
    stop = row[1]['End'] # no -1 necessary since end-point is included in .ptt but excluded in python indexing.
    strand = row[1]['Strand']

    geneCounts = genomeCounts[ strand ][start - offset :stop + offset] # Assign counts to individual genes.

    # To account for genes whose start / end coordinates are less than 'offset' nt away from the ends of the parent sequence.
    if start - offset < 0:
        geneCounts = numpy.pad( genomeCounts[ strand ][:stop + offset], (offset - start,0), mode='constant' )
    if stop + offset >= len( genomeCounts[ strand ]):
        geneCounts = numpy.pad( genomeCounts[ strand ][start - offset : ], (0, offset + stop - len(genomeCounts[ strand ]) ), mode='constant' )
    if start - offset < 0 and stop + offset >= len(rpCounts[ parentSeqStrand ]):
        geneCounts = numpy.pad( genomeCounts[ strand ], ( offset - start, offset + stop - len(genomeCounts[ strand ]) ), mode='constant' )

    seq = genome[ start:stop ].seq    

    if strand == '-': # Invert the counts array and reverse_complement the corresponding sequence.
        geneCounts = geneCounts[::-1]
        seq = seq.reverse_complement()

    meanReadDensity = geneCounts[+offset:-offset].sum() / float( len(seq) )
    genes[ gene ] = dict( [ ('totalReads', geneCounts[offset:-offset].sum()), \
                        ('meanReadDensity', geneCounts[offset:-offset].sum() / (stop-start) ) ] )

    conditions['meanReadDensity'] = ( meanReadDensity > thresholdReadDensity ) # check if mean read density for this gene is above threshold.       
    # if gene does not have thresholdReadDensity, ignore it for analysis.
    if not  conditions['meanReadDensity']:
        continue
        
    frontSensitiveCounts = 0
    midSensitiveCounts = 0    
        
    # Identify sequence features of interest. (typically these are codons).                     
    for featureLocation in range( 0,len(seq),3 ):
        feature = str( seq[featureLocation: featureLocation+3] )
        if featureLocation == 0:
            feature = 'start'
        features[ feature ][ 'featureUsage' ] += meanReadDensity # This counts the usage of the feature in the total mRNA / monosome pool.
        conditions['isFeature'] = ( feature in features ) #check if feature is a codon.        

        # threshold based on whether feature satisfies all conditions.
        if not ( conditions['isFeature'] ):
            continue

        features[ feature ][ 'readDensitySum' ] = features[ feature ][ 'readDensitySum' ] + geneCounts[featureLocation:featureLocation+2*offset+1] / meanReadDensity
        features[ feature ][ 'readDensitySquareSum' ] = features[ feature ][ 'readDensitySquareSum' ] + ( geneCounts[featureLocation:featureLocation+2*offset+1] / meanReadDensity )**2
        features[feature]['featureCounts'] += 1
        
        if feature in ['CTA', 'CTC', 'CTT']:
            if featureLocation in range(300):
                frontSensitiveCounts += 1
            elif featureLocation in range(300,600):
                midSensitiveCounts += 1
                
    minLength = min( [len(seq), len(profileDensity)] )
    profileDensity[:minLength] = profileDensity[:minLength] + geneCounts[+offset:offset+minLength] / meanReadDensity
    profileCounts[:minLength] = profileCounts[:minLength] + 1
    if frontSensitiveCounts <= 1 and midSensitiveCounts <= 1:
        robustProfileDensity[:minLength] = robustProfileDensity[:minLength] + geneCounts[+offset:offset+minLength] / meanReadDensity
        robustProfileCounts[:minLength] = robustProfileCounts[:minLength] + 1
    if frontSensitiveCounts >= 3 and midSensitiveCounts <= 1:
        frontSensitiveProfileDensity[:minLength] = frontSensitiveProfileDensity[:minLength] + geneCounts[+offset:offset+minLength] / meanReadDensity
        frontSensitiveProfileCounts[:minLength] = frontSensitiveProfileCounts[:minLength] + 1
    if frontSensitiveCounts <= 1 and midSensitiveCounts >= 3:
        midSensitiveProfileDensity[:minLength] = midSensitiveProfileDensity[:minLength] + geneCounts[+offset:offset+minLength] / meanReadDensity
        midSensitiveProfileCounts[:minLength] = midSensitiveProfileCounts[:minLength] + 1

    rowCounts += 1

sampleStats = dict()
sampleStats['profileDensity'] = profileDensity
sampleStats['profileCounts'] = profileCounts
sampleStats['robustProfileDensity'] = robustProfileDensity
sampleStats['robustProfileCounts'] = robustProfileCounts
sampleStats['frontSensitiveProfileDensity'] = frontSensitiveProfileDensity
sampleStats['frontSensitiveProfileCounts'] = frontSensitiveProfileCounts
sampleStats['midSensitiveProfileDensity'] = midSensitiveProfileDensity
sampleStats['midSensitiveProfileCounts'] = midSensitiveProfileCounts

pickle.dump( genes, open( 'output_files/sequencing_output/geneStats_'+ sampleName + '.pickled', 'wb' ))
pickle.dump( features, open( 'output_files/sequencing_output/featureStats_' + sampleName + '.pickled', 'wb' ) )
pickle.dump( sampleStats, open( 'output_files/sequencing_output/sampleStats_' + sampleName + '.pickled', 'wb' ) )
```

#### Run analysis ####

```python
%%writefile masterscript.sh
# %%bash

# Trim polyA, align using bowtie and then calculate statistics for each sample. These steps will take ~20 min per sample.
# Can be run on a cluster to speed up. In this case, pass $file as argument to the bash script and set $file=$1 inside.

for file in rawdata/*.sra;
    do
    echo $file;
    fastq-dump -O rawdata $file; # sra to fastq.
    echo "Finished converting sra to fastq.";
    python scripts/polyatrimmer.py ${file/sra/fastq}; # trim polyA.
    echo "Finished trimming polyA tails.";
    scripts/bowtie-0.12.8/bowtie -v 2 nc_000913.3 ${file/sra/trimmed} ${file/sra/genome.aligned}; # align.
    echo "Finished alignment.";
    python scripts/collapsereads.py ${file/sra/genome.aligned}; # aggregate reads at each coordinate.
    echo "Finished collapsing reads.";
    mkdir output_files;
    mkdir output_files/sequencing_output;
    python scripts/calculate_genestats_and_samplestats.py ${file/sra/genome.pickled}; # calculate codon and gene statistics.
    echo "Finished calculating statistics.";
    done;
```

```python
samples = ['wt_noleu_mono','wt_noser_mono','wt_yesser_mono'              
                ]

subsetGenes = ['leuL', 'lptA', 'fabI' ]

offset = 100

genome = Bio.SeqIO.read( 'seq_and_annotation/NC_000913.fna', 'fasta' )
geneTable = pandas.read_table( 'seq_and_annotation/NC_000913.ptt', skiprows=2 )
geneTable['Start'] = geneTable['Location'].apply( lambda x: numpy.int32( x.split('.')[0] ) ) # Parse start coordinate of each CDS.
geneTable['End'] = geneTable['Location'].apply( lambda x: numpy.int32( x.split('.')[-1] ) ) #Part end coordinate of each CDS.

genes = dict()
for row in geneTable.iterrows():
    gene = row[1]['Gene']
    if gene in genes:
        gene = gene + row[1]['Synonym']    
    genes[ gene ] = {
                     'function': row[1]['Product'],
                     'strand':  row[1]['Strand'],
                     'start': row[1]['Start'] - 1,
                     'stop': row[1]['End'],
                     'length': row[1]['End'] - row[1]['Start'] + 1 - 3 # exclude stop codon in length
                     }
    
xmax = max( [genes[gene]['length'] for gene in subsetGenes] ) 

fig = plt.figure()
common_ax = plt.axes(frameon=False)
common_ax.set_xticks([])
common_ax.set_yticks([])
common_ax.yaxis.labelpad = 40
common_ax.xaxis.labelpad = 40
common_ax.set( ylabel = 'Number of Sequencing Reads' )
common_ax.set( xlabel = 'Distance from Start Codon (nt)' )

numberOfAxes = len(samples) * len(subsetGenes); firstplot = True

for sampleIndex, sample in enumerate( samples ):    
    rpData = pickle.load( open( 'output_files/sequencing_output/' + sample + '.genome.pickled' ) )
    for geneIndex, gene in enumerate( subsetGenes ):
        
        geneData = numpy.zeros( (genes[gene]['length'] + 2*offset + 3,), dtype=float )
        for locindex, location in enumerate( range( genes[gene]['start']-offset, genes[gene]['stop']+offset ) ):
            rpDataKey = rpData.keys()[0][:-1] + genes[gene]['strand']
            if location in rpData[ rpDataKey]:
                geneData[locindex] = rpData[ rpDataKey][location]
                
        seq = genome.seq[ genes[gene]['start']:genes[gene]['stop'] ]
        if genes[gene]['strand'] == '-' :
            seq = seq.reverse_complement()
            geneData = geneData[::-1]
        seq = str(seq)                
        
        ax = fig.add_subplot( numberOfAxes,1 , len(samples) * geneIndex + sampleIndex + 1 )
        
        if sample.find('noleu') != -1:
            fillColor = 'grey'
        elif sample.find('noser') != -1:
            fillColor = 'ghostwhite'
        elif sample.find('yes') != -1:
            fillColor = 'lightgrey'
        
        ax.fill_between(numpy.arange(0,len(geneData)-2*offset), geneData[offset:-offset], edgecolor = 'black', facecolor = fillColor );
        
        spineOffset = {'left': 5, 'bottom': 0}    
        [ spine.set_position(('outward',spineOffset[loc])) if loc in ['left','bottom'] else spine.set_color('none') for loc, spine in ax.spines.items() ]         
                
        ax.set_xlim(left=0, right=xmax )
        ax.xaxis.set( ticks_position = 'bottom')            
        ax.set_ylim( top = 1.1*ax.get_ylim()[1], bottom=0 )
        ax.yaxis.set( ticks_position = 'left', major_locator = MaxNLocator(2, prune='upper') )   
        
        ymax = ax.get_ylim()[1]
        ymin = ax.get_ylim()[0]
         
        if (sampleIndex != 2) or (gene != 'fabI'):
            ax.xaxis.set_visible( False )
            [ spine.set_color('none') for loc, spine in ax.spines.items() if loc in ['bottom']  ] 
        
        geneLength = genes[gene]['length']
        
        for location in range(0, len(seq), 3 ):
            codon = seq[location:location+3]
            if ((sample.find('noleu') != -1 and codon in aaCodondict['leu']) 
                                or (sample.find('noser') != -1 and codon in aaCodondict['ser'])):                  
                temp = ax.annotate("",
                xy=(location, -0.02*ymax), xycoords='data',
                xytext=(location, -0.1*ymax), textcoords='data',
                arrowprops=dict(arrowstyle="simple", color=edgeColors[codon]), annotation_clip=False)
                
        if sampleIndex == 0:
            ax.text(geneLength/2, 1.01*ymax, gene, ha='center', fontstyle='italic', fontsize=10)            
        
fig.set_size_inches([4,0.8*numberOfAxes]) 
fig.savefig( 'gene_rp_profile.svg' )
```
