```python
import time
from IPython import display
import pandas # tabular data processing library.
import cPickle as pickle # python native format storage library.
import re # regular expression 
import scipy.ndimage.filters # for smoothing data
import matplotlib.pyplot as plt # plotting library
import Bio.SeqIO # dna sequence input and output.
from matplotlib.ticker import MaxNLocator, LogLocator, AutoMinorLocator # custom tick locators for plots
from IPython.html.widgets import interact # for interactive data analysis
import itertools # for cycling through lists

aaCodondict = {
                'leu' : ['CTA', 'CTC', 'CTG', 'CTT', 'TTA', 'TTG' ],
                'ser' : ['AGC', 'AGT', 'TCA', 'TCC', 'TCG', 'TCT' ],
                }
aaTrnadict = {
                'leu' : ['CAA','CAG','GAG','TAA','TAG'],
                'ser' : ['CGA', 'GCT', 'GGA', 'TGA'],
            }

edgeColors = { 'CTA' : 'dodgerblue',
               'CTC' : 'forestgreen',
               'CTG' : 'darkred',
               'CTT' : 'mediumslateblue',

               'TTA' : 'red',
               'TTG' : 'chocolate',
               
               'TCA' : 'olive',
               'TCC' : 'mediumslateblue',
               'TCG' : 'dodgerblue',
               'TCT' : 'forestgreen',
               'AGC' : 'chocolate',
               'AGT' : 'red',
               }

# mRNA species that have less than these many initations during the simulation are not considered
# for ribosome occupancy measurements.
minimumNumberOfInitations = 5
# This abortion rate is used for all plots where abortion rate is not varied.
defaultAbortionRate = '0.100000' 
# This is the wild-type number of ribosomes in the cell.
defaultNumberOfRibosomes = '44000'
# This is the relative aminoacylation rate for all plots where this value is not varied.
defaultStarvationLevel = '0.010000'
numberOfSimulatedGenes = 1518
```

<a id="fig3a"></a>
#### Figure 3A ####

```python
sequenceFile = open( "seq_and_annotation/Ecoli.input.seq", "read" )

genes = dict()

for lineNumber, line in enumerate( sequenceFile ):
    values = line.split()
    genes[ lineNumber ] = dict()
    genes[ lineNumber ]['initiationRate'] = float( values[0] ) # used as input to the simulation.
    genes[ lineNumber ][ 'mrnaLevel' ] = int( values[1] ) # copies per cell.
    genes[ lineNumber ][ 'codonSequence' ] = numpy.array( [ int(x) for x in values[2:] ], dtype=int )

codons = dict()
codonFile = open( "seq_and_annotation/genetic.code.tsv", "read" )
for codonindex, line in enumerate( codonFile.readlines()[1:] ):
    codons[ codonindex ] = { 'triplet': line.split()[1],}


fig = plt.figure()
fig.set_size_inches((5,5))  
common_ax = plt.axes(frameon=False)
common_ax.set_xticks([])
common_ax.set_yticks([])
common_ax.yaxis.labelpad = 50
common_ax.xaxis.labelpad = 40

axisIndex = 0    

xaxis_file = ["output_files/simulation_v2/output_v2_thresholdtime_1000_totaltime_1500_ribosomes_44000"\
              "_abortthreshold_10_abortion_0.100000_ribotrnakcatkm_2e6_"\
              "minelongtime_0.00_gene_ribo_density.out"]
yaxis_file = ["output_files/simulation_v2/output_v2_thresholdtime_1000_totaltime_1500_ribosomes_44000"\
              "_abortthreshold_10_abortion_0.100000_ribotrnakcatkm_2e7_"\
              "minelongtime_0.05_gene_ribo_density.out"]

sampleIndex = 0
codonDensity = dict()
for File in xaxis_file + yaxis_file:
    outputRibosomeDensityFile = open( File, "read" )
    for lineNumber, line in enumerate( outputRibosomeDensityFile ):
        values = line.split()
        if lineNumber in genes:
            genes[ lineNumber ][ 'ribosomeDensity' ] = numpy.array( [ float(x) for x in values ], dtype=float )

    initiationRateFile = open( File.replace('gene_ribo_density','gene_initimes'), "read" )    
    initiationRateFile.readline() # skip header
    for lineNumber, line in enumerate( initiationRateFile ):
        values = line.split()
        if lineNumber in genes:
            genes[ lineNumber ][ 'numberOfInitiations' ] = int( values[1] )

    for codonindex in codons:
        codons[ codonindex ][ 'ribosomeDensity' ] = 0
        codons[ codonindex ][ 'frequency' ] = 0
        
    for gene in genes:
        if genes[gene]['numberOfInitiations'] < minimumNumberOfInitations:
            continue
        for location, codonindex in enumerate( genes[gene]['codonSequence'] ):
            codons[ codonindex ]['ribosomeDensity'] += ( genes[ gene ][ 'ribosomeDensity' ][location] / genes[gene]['ribosomeDensity'].mean() )
            codons[ codonindex ]['frequency'] += 1

    codonDensity[sampleIndex] = dict()
    for codonindex in codons:
        codonDensity[ sampleIndex ][ codonindex ] = codons[ codonindex ]['ribosomeDensity'] / codons[ codonindex ]['frequency']
    
    sampleIndex += 1
    
codonDensityDf = pandas.DataFrame.from_dict( codonDensity )

ax = fig.add_subplot(2,2,axisIndex+1)
spineOffset = {'left': 5, 'bottom': 0}    
[ spine.set_position(('outward',spineOffset[loc])) if loc in ['left','bottom'] else spine.set_color('none') for loc, spine in ax.spines.items() ]         


ax.loglog( codonDensityDf[0] / codonDensityDf[0].mean(), codonDensityDf[1] / codonDensityDf[1].mean(),
            linestyle = 'none', marker = '.', markersize=4, color='black', markerfacecolor='grey', basex=2, basey=2 )
ax.set_ylim( ax.get_xlim() )
ax.xaxis.set( ticks_position = 'bottom', major_locator = LogLocator(base=2,numticks=8) )
ax.yaxis.set( ticks_position = 'left', major_locator = LogLocator(base=2,numticks=8) )

ax.set_xlabel( u'Ribosome Occupancy (τ=0s)' )
ax.set_ylabel( u'Ribosome density (τ=0.05s)' )

axisIndex += 1;

fig.savefig('fig3a.svg')
```

#### Figure 3B ####

```python
allFiles = !ls output_files/simulation_v3/*_gene_totetimes.out

abortion=defaultAbortionRate
ribosomes=defaultNumberOfRibosomes

fig = plt.figure()
fig.set_size_inches((5,5))  
common_ax = plt.axes(frameon=False)
common_ax.set_xticks([])
common_ax.set_yticks([])
common_ax.yaxis.labelpad = 50
common_ax.xaxis.labelpad = 40

axisIndex = 0    
subsetFiles = filter( lambda x: x.find("ribosomes_" + str(ribosomes) + "_") != -1, allFiles )
subsetFiles = filter( lambda x: x.find("abortion_" + abortion + "_") != -1, subsetFiles )

markerfacecolor = {'0.00': 'white', '0.05': 'grey' }
markerList = {'1': 's', '10': 'o', '50': '^', }

ax = fig.add_subplot(2,2,axisIndex+1)

for codonelongtime in ['0.00','0.05']:
    mrnaFractionValues = numpy.unique( 
                [re.search( 'mrnafraction_([\d\.]+)_', x ).groups(0)[0] for x in subsetFiles ] )    

    for mrnafraction in mrnaFractionValues:
        totalNumberOfProteins = dict()
        cai = dict()
        sampleIndex = 0
        subsubsetFiles = filter( lambda x: x.find( 'minelongtime_' + codonelongtime + '_' ) != -1, subsetFiles )
        subsubsetFiles = filter( lambda x: x.find( 'mrnafraction_' + str(mrnafraction) + '_' ) != -1, subsubsetFiles )
        for File in subsubsetFiles: 
            totalNumberOfProteins[ sampleIndex ] = 0            
            proteinsProduced = 0
            mrnaNumber = 0
            for line in open( File ).readlines()[1:]: 
                geneId = int( line.split()[0] )
                if geneId in range(numberOfSimulatedGenes,numberOfSimulatedGenes+30):
                    proteinsProduced += float( line.split()[1] )
            for line in open( File.replace('totetimes.out','data.in') ).readlines()[1:]: 
                geneId = int( line.split()[0] )
                if geneId in range(numberOfSimulatedGenes,numberOfSimulatedGenes+30):
                    mrnaNumber += float( line.split()[2] )                 

            totalNumberOfProteins[ sampleIndex ] =   proteinsProduced / float( mrnaNumber )

            cai[ sampleIndex ] = float( re.search( 'cai_([\d\.]+)_', File ).groups(0)[0] ) 
            sampleIndex += 1

        caiArray = numpy.zeros( (len(cai),), dtype=float )    

        for index, key in enumerate( sorted( cai, key = cai.get ) ):                
            caiArray[ index ] = cai[ key ]                 

        totalNumberOfProteinsArray = numpy.zeros( (len(totalNumberOfProteins),), dtype=float)    
        for index, key in enumerate( sorted( cai, key = cai.get ) ):
            totalNumberOfProteinsArray[ index ] = totalNumberOfProteins[ key ]

        ax.plot( caiArray, totalNumberOfProteinsArray / totalNumberOfProteinsArray.max(),
                '-', marker = markerList[mrnafraction], label = mrnafraction + "%", markersize=6,
                markerfacecolor = markerfacecolor[codonelongtime], markeredgecolor = 'black', color = 'black' )

spineOffset = {'left': 5, 'bottom': 0}    
[ spine.set_position(('outward',spineOffset[loc])) if loc in ['left','bottom'] else spine.set_color('none') for loc, spine in ax.spines.items() ]         

ax.set_xlim(left = 0, right = 1)
ax.set_ylim(bottom = 0.2, top = 1.05 )
ax.xaxis.set( ticks_position = 'bottom', ticks=numpy.arange(0,1.1,0.2) )
ax.yaxis.set( ticks_position = 'left', ticks=numpy.arange(0.2,1.1,0.2) )
ax.set_xlabel( 'Codon Adaptation Index' )
ax.set_ylabel( 'Transgene proteins / mRNA\n(relative to max)' )    
temp = ax.legend(loc='lower right', title='Transgene fraction\nof transcriptome', bbox_to_anchor=(1.3,0,0.2,0.2) )

fig.savefig('fig3b.svg')
```

```python
allFiles = !ls output_files/simulation_v1/*starvationfactor*gene_totetimes.out

sequenceFile = open( "seq_and_annotation/Ecoli.input.seq", "read" )

lengths=list()
for lineNumber, line in enumerate( sequenceFile ):
    values = line.split()
    lengths.append( len( values[2:] ))

fig = plt.figure()
fig.set_size_inches((5,5))

axisIndex = 0

aminoacylationRateConstant = defaultStarvationLevel
numberOfRibosomes = defaultNumberOfRibosomes

subsetFiles = filter( lambda x: x.find("ribosomes_" +  numberOfRibosomes + "_") != -1, allFiles )
richControl = filter( lambda x: x.find("starvationfactor_" + "1.000000" +"_") != -1, subsetFiles ) 

subsetFiles = filter( lambda x: x.find("starvationfactor_" + aminoacylationRateConstant +"_") != -1, subsetFiles ) 

proteinDf = pandas.DataFrame()
for aminoacid in ['leu','ser']:
    subSubsetFiles = filter( lambda x: x.find( aminoacid ) != -1, subsetFiles + richControl)

    chargedFractions = numpy.unique( 
                [re.search( 'starvationfactor_([\d\.]+)_', x ).groups(0)[0] for x in subSubsetFiles ] )
        
    ax = fig.add_subplot(2,2,axisIndex+1)
    
    spineOffset = {'left': 5, 'bottom': 0}    
    [ spine.set_position(('outward',spineOffset[loc])) if loc in ['left','bottom'] else spine.set_color('none') for loc, spine in ax.spines.items() ]         
    
    ax.set_xlim( left = 2**-11, right = 2**5)

    totalNumberOfProteins = dict()
    proteinDict = dict()
    normalization = 0
    for chargedFractionIndex, chargedFraction in enumerate(chargedFractions[::-1]):
        totalNumberOfProteins[ chargedFractionIndex ] = dict()
        abortionRate = dict()
        sampleIndex = 0
        subsubsetFiles = filter( lambda x: x.find( 'starvationfactor_' + str(chargedFraction) + '_' ) != -1, subSubsetFiles )
        for File in subsubsetFiles: 
            totalNumberOfProteins[ chargedFractionIndex ][ sampleIndex ] = dict()            
            lineNumber = 0
            for line in open( File ).readlines()[1:numberOfSimulatedGenes+1]:
                totalNumberOfProteins[ chargedFractionIndex ][ sampleIndex ][ lineNumber ] = float( line.split()[1] ) * lengths[lineNumber]
                lineNumber += 1

            abortionRate[ sampleIndex ] = float( re.search( 'abortion_([\d\.]+)_', File ).groups(0)[0] ) 
            if abortionRate[ sampleIndex ] == 0:
                abortionRate[ sampleIndex ] = (2.0)**(-11.0)
                
            if '%0.3f'%abortionRate[sampleIndex] in ['0.000', '0.100']:
                proteinDict[ aminoacid + '_' + chargedFraction[:5] + '_%0.3f'%abortionRate[sampleIndex] ] = totalNumberOfProteins[ chargedFractionIndex ][ sampleIndex ]
                
            sampleIndex += 1

        abortionRateArray = numpy.zeros( (len(abortionRate),), dtype=float )    

        for index, key in enumerate( sorted( abortionRate, key = abortionRate.get ) ):
            abortionRateArray[ index ] = abortionRate[ key ]                

        totalNumberOfProteinsArray = numpy.zeros( (len(totalNumberOfProteins[ chargedFractionIndex ]),), dtype=float)    
        for index, key in enumerate( sorted( abortionRate, key = abortionRate.get ) ):
            totalNumberOfProteinsArray[ index ] = sum( [ totalNumberOfProteins[ chargedFractionIndex ][ key ][gene] 
                                                        for gene in totalNumberOfProteins[ chargedFractionIndex ][ key ] ] )
            
        if chargedFraction == '1.000000':
            normalization = totalNumberOfProteinsArray[-1]

        ax.semilogx( abortionRateArray, totalNumberOfProteinsArray / normalization * 100, 
                  '-^', label = chargedFraction[:5], basex=2, color = 'black', markersize=6, )    

    ax.set_xlabel( 'Abortion Rate (s-1)' )
    if not axisIndex:
        ax.set_ylabel( 'Synthesis Rate of\n Complete Proteins (% of Rich)' )        
    if aminoacid == 'leu':
        ax.set_ylim(bottom = 10, top = 40)
    if aminoacid == 'ser':
        ax.set_ylim(bottom = 0, top = 30)
    
    ax.yaxis.set( ticks_position = 'left', major_locator= MaxNLocator(3, prune=None) )   
    ax.xaxis.set( ticks_position = 'bottom', major_locator = LogLocator(base=2, numticks=5) )   

    axisIndex += 1
    
    df = pandas.DataFrame( proteinDict ) 
    proteinDf[aminoacid + '_starv_ratio'] = numpy.divide(df[aminoacid + '_0.010_0.100'], df[aminoacid + '_0.010_0.000'])    
    proteinDf = pandas.concat( [df, proteinDf], join='inner', axis=1)

fig.savefig('fig7a_pss.svg')    
```
